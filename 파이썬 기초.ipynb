# 1. 파이썬 시작하기
## 1-1. 파이썬 설치하기
1) Python: python 웹페이지 -> 다운로드  

2) Jupyter
- 설치: 윈도우 키 -> cmd(관리자 권한) -> pip install jupyter
- 실행: 원하는 폴더에서 -> 빈화면 shift 누른채로 마우스 오른쪽 클릭 -> Powershell -> "jupyter notebook" 입력



## 1-2. 주피터 사용법
1) 주피터 셀
- 코드 : 파이썬 코드 입력

- 마크다운(노트북) : 필기용


2) 마크타운 사용법:
- '#'의 갯수가 많을 수록 글씨의 크기가 작아진다.
- 줄바꿈은 문장이 끝난 후에 'space' 버튼을 두번 누른다.

3) 실행  
① Run 버튼  
② 'ctrl + enter': 오직 실행  
③ 'alt + enter'" 실행 + 코드 추가  
- 실행되면 [ ] 안에 숫자가 뜬다. -> 몇 번째 실행 중인지 나타냄

- [ * ]가 뜨면 실행 중 임을 의미한다.
