# 함수(function, fn): 특정 task를 실행하기 위해 만들어진 명령어 블럭(덩어리)

- 함수는 괄호 안에서부터 밖으로 실행된다. ex) print(int("입력하세요:"))
- 점으로 된 함수는 왼쪽에서부터 오른쪽으로 실행된다. ex)

### 1. round(실수, 몇 번째 자리까지 반올림할 것인가): 반올림을 해준다.
### 2. trunc(변수): 내림/ ceil(변수): 올림 (가장 가까운 정수 중에 큰 수) -> ceil()는 자주 사용
### * import 모듈 이름: 파이썬에서 기본적으로 탑재되지 않은 것은 입력하여 실행해야 한다.
- 특정 모듈에서 특정 함수를 불러올 때, "모듈 이름. 함수"

### 3. floor(변수): 가장 가까운 정수 중에 작은 수 (진행 방식은 trunc와 다르나 양수 결과 값은 같음, but 음수에서 다름)
### 4. 변수.lower(): 전부 소문자/ 변수.upper(): 전부 대문자
### 5. 변수.strip(): 양쪽의 공백을 제거해준다. (문장 사이의 공백은 제거하지 X)
### 6. 변수.replace("바꿔줄 대상", "새 문자"): 변수의 일부를 바꿔준다.
- 문장 사이의 공백은 '변수.replace(" ", "")'으로 공백을 없앤다.

### 7. 변수.split(): 나누다, 쪼개다 -> 어떤 기준을 중심으로 문자열을 나누어 준다.
### 8. len(입력): 요소의 길이(length), 즉 갯수를 알려준다.

### 각각의 것들은 문법으로 개발자가 지정해 둔 것이기 때문에 특정한 이유 없이 외운다.

```python
# floor(): 가장 가까운 정수 중에 작은 수
import math

no1 = -31.3

print(math.trunc(no1))
print(math.floor(no1))
print(math.ceil(no1))

# trunc()와 floor()는 양수에서 결과 값이 같으나 음수에서 결과값이 다르다. trunc(음수) > floor(음수)
```

```python
# lower(): 전부 소문자/ upper(): 전부 대문자

str2 = "PyThoN"

print(str2.lower())
print(str2.upper())
str2.upper().upper()
# 점으로 된 값들은 왼쪽에서 오른쪽으로 출력
```

```python
# strip(): 공백을 제거해 준다. -> 문장 사이의 공백은 제거할 수 없다.
# 문법: 변수.strip()
str3 = "        공 백으로 가득 찬 문     장        "
str3.strip() # 양쪽의 공백을 없애준다.
str3.lstrip() # 왼쪽의 공백만 없애준다.
str3.rstrip() # 오른쪽의 공백만 없애준다.
# 앞뒤로 공백이 있는 경우 원인 없이 Error가 발생할 수 있기 때문에 strip을 이용하여 보이지 않는 공백을 제거해 준다.
```

# 문자열 (String)
- 문자열을 표현하기 위해서는 " " 또는 ' '
- index

1. 정방향일 때 왼쪽 끝 0에서 오른쪽으로 증가한다. (실제 위치에서 -1을 하면 됨)
2. 역방향일 때 오른쪽 끝 -1에서 왼쪽으로 감소한다. 
- 범위를 정해줄 때: str[시작:끝] 시작 값은 포함(inclusive), 끝 값은 불포함 -> '시작:끝'의 의미는 '시작<->끝'

```python
# 여러개를 입력해도 모든 output을 출력해준다.
from IPython.core.interactiveshell import InteractiveShell
InteractiveShell.ast_node_interactivity = "all"
```

```python
str = "파이썬완전쉽죠?"

# index 값은 0부터 시작한다.
str[1] #'이'
str[4] # '전'

# 범위를 정해줄 때: 시작 값은 포함(inclusive), 끝 값은 불포함(declusive)
str[1:4] # '이썬완' 출력
str[1:] # 1부터 마지막까지 전부 의미 '이썬완전쉽죠?'
str[:4] # 4의 앞 전부 의미 '파이썬완'
```
