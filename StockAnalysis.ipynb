```python
# 모든 모듈 열기
from bs4 import BeautifulSoup # 리뷰할 내용 긁어오기
from selenium import webdriver # 웹페이지 열기
import time # 시간(현재/멈춤 등)과 관련된 코드 실행 
import math # 수학적 계산 실행
import os # 파일 생성/열기/지정
import pandas as pd  # 데이터 조작/분석
import numpy as np # 수치계산

# 데이트 캠퍼스 창 열기
from selenium import webdriver
from webdriver_manager.chrome import ChromeDriverManager
driver = webdriver.Chrome(ChromeDriverManager().install())
url = 'https://finance.naver.com/world/sise.naver?symbol=SPI@SPX'
driver.get(url)
driver.maximize_window()
time.sleep(1)

# 쿠키 삭제하기
driver.delete_all_cookies()
time.sleep(1)

# 크롤링할 테이블 지정하기
full_html = driver.page_source
soup = BeautifulSoup(full_html,'html.parser')
total_r = soup.find('table',id="dayTable")
td1=total_r.find_all('td','tb_td')
td2=total_r.find_all('td','tb_td2')
td3=total_r.find_all('td','tb_td3')
td4=total_r.find_all('td','tb_td4')
td5=total_r.find_all('td','tb_td5')
td6=total_r.find_all('td','tb_td6')

td1_new=[]
for i in td1:
    new=i.text.strip()
    td1_new.append(new)
td2_new=[]
for i in td2:
    new=i.text.strip()
    td2_new.append(new)
td3_new=[]
for i in td3:
    new=i.text.strip()
    td3_new.append(new)
td4_new=[]
for i in td4:
    new=i.text.strip()
    td4_new.append(new)
td5_new=[]
for i in td5:
    new=i.text.strip()
    td5_new.append(new)
td6_new=[]
for i in td6:
    new=i.text.strip()
    td6_new.append(new)
    
import pandas as pd
data=pd.DataFrame({
    '일자':td1_new,'종가':td2_new,'전일대비':td3_new,'시가':td4_new,'고가':td5_new,'저가':td6_new
})
data
```
