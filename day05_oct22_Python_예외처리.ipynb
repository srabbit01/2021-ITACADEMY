# 예외처리***

- 예외 = Error (파이썬 Error 종류: https://www.programiz.com/python-programming/exceptions) -> 나중에 Error type만 알면 됨
- 예외처리: Error가 발생했을 경우 어떤 방식으로 해결할지 야기하는 것
- 예외처리를 하는 이유: 크롤링

이때, 2번 과정에서 Error가 나면 저장하기 전에 프로그램이 멈춘다. (이런 경우, data가 다 날아갈 위험이 있음) 따라서, 다른 데이터도 긁어서 저장하기 위해 Error 처리를 해주는 것이다.
- 문법:


```python
print("정상 작동할 경우")
no1 = int(input("숫자를 입력하세요: "))
print(no1)
```

```python
# 위와 같이 숫자를 입력하지 않으면 Error가 나는데, 그것을 방지하기 위해 예외처리
try:
    no1 = int(input("숫자를 입력하세요: "))
except:
    print("'숫자'를 입력하셔야 합니다.")
```

```python
try:
    no1 = int(input("숫자를 입력하세요: "))
    no2 = 100/no1
except ValueError: # Error의 타입을 정해주면 정해진 Error 타입에만 반응함
    print("'숫자'를 입력하셔야 합니다.")
else:
    print("100/%s은 %s입니다." %(no1,no2))
```

```python
try:
    no1 = int(input("숫자를 입력하세요: "))
    no2 = 100/no1
except ValueError: # Error의 타입을 정해주면 정해진 Error 타입에만 반응함
    print("'숫자'를 입력하셔야 합니다.")
except ZeroDivisionError: # Data가 없는 경우(0), 발생할 수 있는 Error
    print("0은 입력하실 수 없습니다.")
except:
    print("unknown error") # Syntax Error는 프로그램이 실행되기 전에 발생하는 Error기 때문에 try 이전에 Error 발생
else:
    print("100/%s은 %s입니다." %(no1,no2))
```

이때, 사용자에게 출력할 건수를 입력 받아서 입력 받은 수만큼 앞에서부터 출력하세요.

요소 수보다 많은 입력(IndexError)이 발생할 경우, '모든 행의 출력 완료되었습니다'라고 표시하는 예외처리를 하라.

```python
try:
    name = []
    no = 0
    num = int(input("몇 개를 출력하시겠습니까? "))
    while not num == 0:
        name.append(input("%s번째 이름을 입력하십시오: " %(no+1)))
        no += 1
        if num == no:
            break
except IndexError:
    print("모든 행의 출력이 완료되었습니다.")
else:
    print("출력을 완료했습니다.")
```

```python
이름 = ["홍길동", "신사임당", "일지매", "전우치", "강감찬"]
no = int(input("몇 번 출력하시겠습니까"))

try:
    for i in range(0,no):
        print(이름[i])
        
except IndexError:
    print("모든 행의 출력이 완료되었습니다.")
else:
    print("출력을 완료했습니다.")
```
