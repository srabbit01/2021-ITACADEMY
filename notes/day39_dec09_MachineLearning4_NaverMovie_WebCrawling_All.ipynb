## TRAIN and TEST DATA 긁어오기

```python
# 모든 모듈 열기
from bs4 import BeautifulSoup # 리뷰할 내용 긁어오기
from selenium import webdriver # 웹페이지 열기
import time # 시간(현재/멈춤 등)과 관련된 코드 실행 
import math # 수학적 계산 실행
import os # 파일 생성/열기/지정
import pandas as pd  # 데이터 조작/분석
import numpy as np # 수치계산

# 제목
print("영화 리뷰 긁어오기")
print()

# 긁어올 리뷰 갯수 정하기
cnt = int(input("긁어올 리뷰 갯수를 적으시오: "))

# 네이버 영화 창 열기
from selenium import webdriver
chrome_path = 'E:\\py_temp\\WebCrawling\\chromedriver.exe'
driver = webdriver.Chrome(chrome_path)
url = 'https://movie.naver.com/'
driver.get(url)
driver.maximize_window()
import time
time.sleep(2)

# 쿠키 삭제하기
driver.delete_all_cookies()
time.sleep(1)

# 입력받은 영화 검색하기 들어가기
review_bar = driver.find_element_by_xpath('//*[@id="scrollbar"]/div[1]/div/div/ul/li[4]/a')
review_bar.click()

# 파일 생성 및 저장할 위치 지정하기
folder_path = 'E:\\py_temp\\최종_제출\\train_review'
if not os.path.isdir(folder_path):
    os.makedirs(folder_path)
os.chdir(folder_path)

# 현재 시간 및 이름 지정하기
# 현재시간
now = time.localtime()
s = '%04d-%02d-%02d-%02d-%02d-%02d' % (now.tm_year, now.tm_mon, now.tm_mday, now.tm_hour, now.tm_min, now.tm_sec)
# 이름 지정
csv_name = "NaverMovie"+"_"+"train_review"+"_"+s+".csv"
```

```python
# 리뷰할 페이지 들어간 후, 크롤링할 리뷰의 갯수 파악하기
full_html = driver.page_source
soup = BeautifulSoup(full_html,'html.parser')
total_r = soup.find('h5','sub_tlt underline').find('strong','c_88 fs_11').text.strip()
total_r = total_r.replace(",","")
total_r = int(total_r)
print("총 %s 건의 리뷰가 있습니다." %total_r)
print()
```

```python
os.chdir(folder_path) # 파일을 저장할 경로 지정

# 데이터 프레임 형태로 리스트 취합하기
movie_frame = pd.DataFrame()
movie_frame['Reviewer Rating']=pd.Series(s_review)
movie_frame['Review']=pd.Series(r_review)

# csv 형태로 저장하기
movie_frame.to_csv(csv_name,encoding="utf-8-sig",index=True)
```
