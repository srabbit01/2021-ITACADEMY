## 목록 스타일
1. 블릿/번호 스타일 지정: list-style-type


- 숫자가 붙는 목록: <ol'> 사용
- 도형 등 숫자가 없는 목록: <ul'> 사용
2. 불릿/번호 대신 이미지 쓰기: list-style-image


3. 목록 들여쓰기: list-style-position

- inside: (기본 위치 기준) 더 안쪽으로 들여쓰기
- outside: 기본 값
4. 목록 속성 한번에 표시하기: list-style

- 특별한 규칙이 없어서 사용하기 용이하다.
- list-style-type와 list-style-image가 충돌할 때, 'list-style-type < list-style-image'


```python
# list-style-type
list-style-type: disc|cicle|...
    
# list-style-image
list-style-image: url(이미지_파일_경로)|none
    
# list-style-position
list-style-position: inside | outside
    
# list-style
list-style: <list-style-type> <list-style-image> <list-style-position>
```

## 표 스타일
1. 표 제목 위치 지정: caption-side

- top: 기본 값
- bottom: 표 아래 쪽에 제목 위치 이동
2. 표 테두리: border


- none은 테두리가 아예 없는 것이고, hidden은 테두리가 있는데 숨긴 것이다.
3. 셀 사이 여백: border-spacing

4. 셀 테두리 합치기: border-collapse

- seperate: 기본 값 (따로 표현하기)
- collapse: 셀 테두리 합치기

```python
# caption-side
caption-side: top|bottom

# border
border: 두께(px) 선_종류(type) 색상

# border-spacing
border-spacing: 수평_거리 수직_거리
    
# border-collapse
border-collapse: collapse|separate
```

## 박스 모델 **
- 대부분의 배치는 박스 안에 들어있다. -> 레이아웃을 잡는데 중요한 역할을 한다.
- 박스모델은 ① 컨텐츠 영역, ② 테두리, ③ 여백 으로 구성된다.

1. 블록 레벨/ 인라인 레벨 요소
- 블록 레벨: <hn'> <p'> <div'> 등등 ① 태그를 사용해서 삽입한 요소가 ② 혼자서 한 줄을 차지할 때 (요소의 너비 = 100%) -> 왼쪽이나 오른쪽에 요소 추가 불가능

- 인라인 레벨: 포함하는 컨텐츠 만큼의 영역만을 차지한다. -> 대개 span을 이용하여 지정

2. 박스 모델의 기본 구성

- content box: html + css 내용이 담긴 것
- padding: 내용(content)과 테두리(border) 사이의 여백
- border: 테두리
- margin: 테두리(border) 밖의 여백
3. width/height 속성: 콘텐츠 영역의 크기 지정
- 크기값: px(절대적), em(절대적), %(상대적), auto(기본값)
4. 박스 모델의 크기 계산: box-sizing  
① content-box: 기본값으로 너비와 높이의 속성이 content 영역만 포함하고 이외는 포함하지 않는다. 예를 들어, .box {width: 350px; border: 10px solid black;}에서 요소의 너비는 370px이다. 즉, 요소의 너비 = content 너비, 높이 = content 높이로 계산된다. (표는 content임을 유의)  
② border-box: 너비와 높이 속성이 테두리까지 포함하고 바깥 여백(margin)은 포함하지 않는다. (요소 상자 안에 안쪽 여백과 테두리 위치) 예를 들어, .box {width: 350px; border: 10px solid black;}에서 요소의 너비는 350px이다. 즉, 요소의 너비 = 테두리 + 안쪽 여백 + content 너비, 높이 = 테두리 + 안쪽 여백 + content 높이로 계산된다.  

- content-box 기준으로 크기를 설정하면 전체박스 크기를 따로 계산해줘야 하는데 이것이 복잡하기에, 레이아웃을 만들 때 대부분의 경우 border-box를 기준으로 크기 설정해 준다.
- 박스를 만들기 위해 되도록 width/height/padding/border를 모두 지정해 주는 것이 좋다. (입력하지 않을 시, padding과 border의 경우 기본 값인 0으로 지정)
5. 박스 모델의 그림자: box-shadow

- 조건: 수평거리와 수직거리는 반드시 지정해 주어야 한다.
- inset: 그림자가 안쪽으로 생긴다.

```python
# box-sizing
box-sizing: content-box|border-box

# box-shadow
box-shadow: 수평거리 수직거리 흐림정도 번짐정도 색상 (inset: 그림자가 밖에 아닌 안으로 생기는 것)
```

## 박스 테두리 스타일
- 테두리 option 값 지정 방향: 시계 방향(top-right-bottom-left)
1. 테두리 스타일 지정: border-style


2. 테두리 두께 지정: border-width

- 테두리의 두계는 상/하/좌/우 각각 설정이 가능하다.
  
① 옵션 1개: 4면에 동일한 옵션 적용  
② 옵션 2개: 옵션1=상/하, 옵션2=좌/우  
③ 옵션 3개: 상/우/하 순서대로 적용한 뒤 좌측은 우측과 동일한 옵션 적용  
④ 옵션 4개 : 상/우/하/좌 시계방향 순서대로 적용  
3. 테두리 색상 지정: border-color

- border-color: 전체 테두리 색상 지정
- border-위치(top/bottom/left/right)-color: 바꾸고 싶은 지점의 테두리 색상 지정

4. 테두리의 모서리 모양(둥글게): border-radius


- %: 테두리를 기준 (%)비율로 지정
- radius 값을 높인다 하여도 최대(max)로 모서리를 지정할 수 있는 한계가 존재한다. (그 이상 둥글게 되지 X)

- 각 꼭짓점을 따로 설정할 수 있다: border-위치-radius

- 선택적으로 "/"한 이후에 한 개에서 네 개의 길이 혹은 % 값을 사용해 추가 반지름을 설정하면 타원형 꼭짓점을 만들 수 있다.

```python
# border-style
border-style: 두께(px) 선_종류(type) 색상
    
# border-width
border-width: 크기|thin|medium|thick

# border-color
border-color: 색상

# border-radius
border-radius: 크기(px,em,%)
```
