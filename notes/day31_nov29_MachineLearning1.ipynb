# Machine Learning
1. 정의
- 데이터에서 지식/ 정보를 자동으로 추출하는 작업
- 스스로 학습하고 배울 수 있는 능력
- 인간이 입력을 통해 지시하지 않고도 기계가 스스로 (발견한 규칙을 이용하여) 결과를 도출하는 것
- 인공지능의 한 분야

2. 인공지능 > 머신러닝 > 딥러닝

3. 일반 머신러닝  
: 일반 머신러닝 알고리즘보다 신경망 알고리즘의 성능이 좋지만 여전히 일반 머신러닝이 더 많이 활용된다.
- 일반 알고리즘은 운용 방법이 쉽다.
- 작은 데이터셋에서도 좋은 성능을 보인다.
- 모델 훈련에 필요한 비용이 적다.
- 결과가 나온 이유를 파악하기 용이하다.
- 결과의 시각화나 해석이 직관적이다.
4. 머신러닝이 필요한 이유: 데이터의 패텬을 스스로 찾아낸다.

5. 머신러닝과 TEP  
: 컴퓨터 프로그램이 특정한 목적(Task)을 달성하기 위해 경험(experience)을 기반으로 행동하고 성능을 평가(Performance) 받으면서 학습해 나가는 것
6. 지도학습 VS 비지도학습  
: 머신러닝은 분석하고자 하는 데이터의 정답(라벨/클래스)이 존재하느냐 없느냐에 따라서 분류가 나누어진다.

7. 강화학습
- 이는 '알파고'로도 잘 알려져 있다.
- 특정 목표에 도달하는 과정에서 보상을 최대화하기 위해 상황에서 취해야 할 조치를 스스로 파악한다.
- 모델 자체가 시행작오를 거치면서 환경에서 적응해 나가는 학습법으로, 지도와 비지도 학습과는 완전히 다른 접근방식이다.

## 머신러닝의 3요소
1. 머신러닝 모델: 코드로 이루어진 블록
- 여러가지 방법을 가공하여 다양한 문제를 해결하는데 활용
- 모델: 굉장히 광범위하고 일반적인 프로그램(코드 블록)에 어떤 데이터를 주입하느냐, 어떤 훈련 과정을 거치느냐에 따라 구체화 된다.
- 예시

: 직선과 실제 판매량의 차는 오차율이라고 하며, 이것이 적어야 정확도가 증가한다.
2. 모델 훈련 알고리즘: 모델을 가지고 있는 데이터를 활용한 트레이닝 >> 세부적인 목적 수행
- 적용시킬 변화를 생각
- 모델에 적용
3. 모델 추론 알고리즘: 훈련된 모델 사용
- 지금까지 처리한 데이터를 활용하여 훈련시킨 모델을 이용해서 각종 예측 작업을 수행하는 것이다.

## 머신러닝 5단계

: 이 모든 단계는 반복적이며, 어느 단계이든 이전 단계로 돌아가 다시 시작할 수 있다.

### 1단계: Task의 정의 및 설정
- 분석을 시작하기 전에 가장 먼저 해야하는 질문: 무엇을 어떻게 해야 할까?
- 문석의 목표:  

  ① WHAT: 할당 받은 과제가 정확하게 무엇인지?  
  - 머신러닝 프로젝트를 진행하기 전에 분석하고자 하는 대상이 무엇인지 판단하고 정의해야 한다.
  - 가능한 세부적이고 "명확한 Task"를 설정해야 한다.
  - 설정해야 하는 Task:  
명확성: 분석하고자 하는 대상이 명확해야 한다.  
세부성 매우 디테일해야 한다.  

  ② HOW: 그리고 그 과제를 어떻게 수행할 것인지?  
  - 데이터 유무: 분석이 가능한 데이터가 있는지, 없으면 만들어 낼 수 있는지, 또 해당 데이터에서 원하는 정답을 찾을 수 있을지?
  - 데이터 형태: 가용한 데이터는 어떤 종류가 있는가?
  - 데이터 평가 방안: 설정한 Task에 대한 평가는 어떻게 할 것인가? 시험 문제에 대한 답안은 존재하는가?
  
- Task는 반드시 디테일하고, 명확하게 설정해야 한다.
- 분석하게 될 데이터에 정답 닶이 있으면 지도학습, 정답값이 없으면 비지도 학습을 실시해야 한다.



#### 지도학습 VS 비지도학습  
: 머신러닝은 분석하고자 하는 데이터의 정답(라벨/클래스)이 존재하느냐 없느냐에 따라서 분류가 나누어진다.

1. 지도 학습 (Supervised Learning)
- 정답이 있는 데이터를 사용해서 Task를 해결하고자 한다면 지도학습 Task를 하고 있는 것
- 정답/라벨/클라스 등 수많은 분야에서 사람들이 모여 만들어 가는 분야인 만큼 다양한 이름들이 있다.
- 현재는 알기 쉽게 '문제'와 '정답'이라는 이름으로 시작한다.

- Labeled Data

- 분석 종류  
① 회귀 분석: 정답 데이터의 종류가 연속형 데이터  
② 분류 분석: 정답 데이터의 종류가 범주형 데이터  
- 정답의 형태

2. 비지도 학습 (Unsupervised Learning)
- 지도 학습과는 반대로 정답이 없는 데이터(Unlabeled Data)를 사용하여 분석을 할 경우 비지도학습 모델을 사용한다. -> 거꾸로 생각하면 비지도학습 모델들은 라벨값이 없더라도 분석을 실시할 수 있다는 뜻
- 사진 등 데이터에서 차이점은 인식할 수 있으나 라벨이 무엇인지 인식할 수 없을때, 분류작업을 실시할 수 있다. 이런 비지도 분류작업을 "군집 분석"이라고 한다.  
예) 
: 두 사진을 비교했을 때, '꽃'이란 라벨링을 인식할 수 없으나 사진 속 꽃의 유무는 파악할 수 있다.

### 3단계: 모델 훈련
: 정제된 데이터를 알고리즘에 입력하여 인공지능 알고리즘을 훈련시키는 것

1. 1단계: 알고리즘 선택하기
- 정답 = 라벨 = 클래스 = 출력변수 = 예측하고자 하는 값 
- 정답의 유무: 분석하고자 하는 데이터에 정답이 존재하는가?  
① 지도(정답O): 가지고 있는 정답 변수의 평태를 살펴보자  

② 비지도(정답X): 군집 분석 실시  
- 정답의 유형  
① 연속형
  - 양의 크기를 나타내기 위해 수량으로 표시되는 변수
  - 주어진 범위 내에서 소수점을 포함해 어떤 값도 가질 수 있다.  
  
  ② 범주형
  - 동일한 성질이나 범위로 나누어진 변수
  - 분류 척도에 따라 관찰한 변수를 가르킨다.
- 지도학습 분류

- 정답의 형태: 연속형 VS 이산형


2. 2단계: 데이터를 훈련 세트와 검정 세트로 나누기
- 모델 훈련 =  학습

: 기계 학습은 존재하는 데이터셋을 훈련세트(공부용 연습문제)와 검정 세트(정답을 제외한 모의고사 문제지=검사용)로 나누어 알고리즘을 학습하는데 사용한다.
- 기본적으로 훈련 데이터셋과 검정 데이터셋은 7:3 ~ 8:2의 비율로 나누어 준다.
- 이를 자동으로 나누어 주는 함수인 train_test_split 함수의 경우 7.5:2.5 비율을 기본값으로 한다.
- 데이터셋을 나누어주는 이유


3. 3단계: 모델 훈련하기
- 모델에 준비된 데이터를 삽입하여 결과에 대한 손실함수(Loss Function)을 계산하고, 손실을 줄여나가는 방향으로 모델 매개변수들을 업데이트하는 것 -> 이때, 사전 정의된 정지 조건에 도달할 때까지 반복
- Loss Function(손실함수): 목표와 모델의 거리를 코드화하는데 사용한다. 즉, 예측 값과 실제값 사이의 평균 거리를 표현한 함수이다. -> 검증 시 오차가 발생하는 패턴을 수학적으로 표현한 함수이다.

4. 4단계: 세부 파라미터 조정하기
- Model Parameters:
  - 모델 파라미터는 모델의 동작 방식을 정의해주는 설정 혹은 구성
  - 상황에 따라 가중치, 편향 등 더 구체적인 이름으로 불리기도 한다.
  - 학습 중에 변경되지 않지만 모델이 학습하는 속도 또는 신회성에 영향을 줄 수 있다.
  - 손실을 최소화시켜주는 파라미터를 설정해야 한다.
- 모델 훈련 시 TIPS:  
① 기존에 만들어진 알고리즘 활용: 대부분 이미 만들어진 것으로 해결 가능  
② "Model Selection" 과정 거치기: 모델 선택 과정은 머신러닝 알고리즘을 여러 후보군으로 설정하여 최종적으로 사용할 알고리즘으로 줄여나가는 과정을 의미  
③ 모델 별 하이퍼파라메터(Hyperparameter)들을 익히자: 세부 변수들은 모델을 훈련할 때는 조정이 불가능하나, 모델이 얼마나 빠르고 안정적으로 훈련되는지에 영향을 줄 수 있다.  
④ 어느 단계든 전단계로 돌아가 분석과정을 반복(iteration)할 준비하자: 머신러닝은 시행착오를 거쳐 조정해야 하는 분야이다.

## 적합성과 일반화
- 머신러닝에서는 실전적용 > 검정세트 > 훈련세트 순으로 중요도가 나뉜다.
- 학습 결과를 두고 아예 학습을 하지 못한 경우에는 '과소적합'이, 검증세트만 망치는 경우에는 과대적합이, 둘 다 잘한 경우에는 일반화가 잘되었다 한다.  
① 과소적합: 예측선이 잘 그어지지 않아(학습이 되지 않아) 예측 성능이 떨어진 것 -> 미지의 데이터를 주더라도 예측하지 못할 확율이 높다.

② 과대적합: 훈련세트를 너무 많이 한 나머지 예측선이 일반화의 범주를 벗어난 상황 -> 머신러닝의 숙적: 학습 데이터는 맞출 수 있으나 새로운 데이터 입력 시 엉뚱한 값을 계산할 수 있다.

③ 일반화

- 적합도 문제 탐지하기: 지금 해당 모델은 데이터를 제대로 학습하고 있을까?

- 손실(Loss): 모델이 학습하면서 예측을 했을 때 발생하는 오차
  - 학습이 진행되면서 손실값(Training Loss)이 점차 0을 향해 수렴해 간다면 학습이 제대로 진행되고 있다고 볼 수 있음 -> 그러나 일반화 된다고 볼 수 X
  - 가장 중요한 것은 연습문제(Training Loss)보다 모의고사 성적(Validation Loss)
  
  - Validation Loss 값이 지속적으로 우하향하더라도 Training Loss와 Validation Loss 값이 너무 벌어지기만 하면 학습이 제대로 진행되었다고 할 수 X
