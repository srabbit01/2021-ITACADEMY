## 서브쿼리(하위쿼리) & ANY/ ALL / SOME
1. 서브쿼리: 쿼리문 안에 들어가는 쿼리문  
ex) userTBL에서 '성시경보다 키가 큰 사람의 이름(name)과 키(height)를 출력하라.

2. ANY: 조건문의 OR 과 같은 역할을 한다.  
ex) userTBL에서 대구인 사람보다 키가 큰 사람을 출력하라. (대구인 사람은 '김경호', '김태희' 두명)

3. ALL: 조건문의 AND 과 같은 역할을 한다. 즉, 주어진 조건 모두를 만족해야 출력이 됨을 의미한다.
4. SOME = ANY와 같은 역할을 한다.

## SQL의 종류
1. DML(Data Manipulation Language):
- 데이터 조작 언어로, '존재하는' 데이터를 조작하는데 사용하는 언어이다.
- 이를 쓰기 위해서는 이미 Table이 정의되어 있어야 한다.
- SELECT/INSERT/DELETE/UPDATE 등
2. DDL:
- 데이터 정의 언어로, 개체(테이블,뷰,인덱스 등)를 생성/삭제/변경하는 언어이다.
- CREATE/DROP/ALTER 등
3. DCL(Data Control Language):
- 사용자에게 권한을 주고 빼았는다.
- GRAND: 권한을 주는 것
- REVOKE: 권한을 빼앗는 것
- DENY: 권한 부여를 거절하는 것
