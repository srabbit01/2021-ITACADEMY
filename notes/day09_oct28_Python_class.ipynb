# Object Oriented Programming

### 1. Procedural Programming (절차지향 프로그래밍)
- 데이터를 중심으로 함수를 구현한다.
- 지금까지는 순차적으로 진행되는 Procedural Programming을 해왔다. 현재 대표적으로 'C'언어에서만 사용된다.
- 초점: '실행' -> 함수 중심
- 지향점: 문제를 해결하기 위한 논리/흐름 모델링
- 장점: Flow가 컴퓨터 처리 구조와 유사하기 때문에 처리 속도가 매우 빠르다.
- 단점:  
  ① 복잡해질 수록 유지/보수가 어렵다.  
  ② 실행 순서가 정해져 있다. -> 실행 순서가 동일하지 않으면 결과가 바뀔 수도 있음  
  ③ Debugging이 어렵다.  

### 2. Object Oriented Programming (객체지향 프로그래밍)
- 기능을 중심으로 매서드를 구현한다.
- 실제 세계를 추상화라는 방법을 이용하여 간단화하는 것이다.
- 초점: '관계도' -> 기능 중심
- 키워드(Ketword):  
  ① 묶음: 필요한 모듈끼리 묶어두는 것 -> 블럭 단위로 되어있다.  
  ② 상속
- 지향점: 실제 세계 모델링
- 장점:  
  ① 코드의 재활용성이 높다.  
  ② 코딩이 상대적으로 간편하다.  
  ③ Debugging이 쉬워진다.  
- 단점:  
  ① 처리속도가 느리다.  
  ② 설계에 비용(시간/돈)이 많이 든다.  
- 단점에 비해 장점이 매우 크기 때문에, 객체지향 프로그래밍이 많이 이용된다.

# Class

- 클래스 이름은 무엇을 하는지, 어떤 객체가 사용하는지 등 정보를 담아준다. -> 단어 앞글자가 대문자인 Camel 표기법을 이용 ex) CreateStudent
- 객체 생성자 = __init__(self,...): 이는 데이터(객체)를 불러올 때 사용된다.

```python
students = [
    Student("윤인성",97,98,88,95),
    Student("연하진",92,98,96,98),
    Student("구지연",76,96,94,90),
    Student("나선주",98,92,96,92),
    Student("윤아린",95,98,98,98),
    Student("윤명월",64,88,92,92)
]

# Class의 선언
class Student:
    # 시작 전, 틀 만들기 -> 누구냐? = 어떤 정보를 가지고 있는가?
    def __init__(self,name,korean,math,english,science): # initionalizer/constructor(생성자): class를 시작할때 넣는 암묵적인 규칙
        self.name = name
        self.korean = korean
        self.math = math
        self.engligh = engligh
        self.science = science
    
    # class가 무엇을 하는가? Method = class 안에 사용자정의함수
    def get_sum(self):
        return self.korean + self.math + self.engligh + self.science
    
    def get_average(self):
        return self.get_sum() / 4 # get_sum안에 있는 함수를 실행함을 의미
    
    def to_string(self):
        return "{}\t{}\t{}".format(
            self.name,
            self.get_sum(),
            self.get_average())
    
print("이름", "총점", "평균", sep="\t")
for i in students:
    print(i.to_string())
```

- Bonus: 계산기 사용현황을 출력하는 함수를 포함한다.  
  ex) cal.ShowCount() 입력시 아래와 같이 출력:  
  덧셈 : 1  
  뺄셈 : 1  
  곱셈 : 2  
  나눗셈 : 0

```python
class MathCalculate:
    def __init__(self,num1,ask,num2):
        self.num1 = num1
        self.ask = ask
        self.num2 = num2
    def calculate(self):
        if self.ask == "+":
            num3 = self.num1+self.num2
            print("%s + %s = %s" %(self.num1,self.num2,num3))
            self.num1 = num3
        elif self.ask == "-":
            num3 = self.num1-self.num2
            print("%s - %s = %s" %(self.num1,self.num2,num3))
            self.num1 = num3
        elif self.ask == "*":
            num3 = self.num1*self.num2
            print("%s * %s = %s" %(self.num1,self.num2,num3))
            self.num1 = num3
        else:
            num3 = self.num1/self.num2
            print("%s / %s = %s" %(self.num1,self.num2,num3))
            self.num1 = num3
    def get_result(self):
        self.calculate()
        print()
        want = input("계속 계산하시겠습니까?(yes/no )").lower()
        while want == "yes":
            self.ask = input("어떻게 연산하시겠습니까?(+,-,*,/) ")
            self.num2 = int(input("계산할 숫자를 추가해주세요: "))
            self.calculate()
            print()
            want = input("계속 계산하시겠습니까?(yes/no ) ").lower()
        if want == "no":
            print()
            return f"최종 계산값은 {self.num1}입니다."
            
num1 = int(input("숫자를 입력하세요: "))
ask = input("어떻게 연산하시겠습니까?(+,-,*,/) ")
num2 = int(input("계산할 숫자를 추가해주세요: "))            
c = MathCalculate(num1,ask,num2)
print(c.get_result())
```

```python
# Another Method
class Calculator :
    def __init__(self):
        self.AddCount = 0
        self.SubCount = 0
        self.MulCount = 0
        self.DivCount = 0

    def Add (self, num1, num2):
        self.AddCount += 1
        return num1 + num2

    def Sub (self, num1, num2):
        self.SubCount += 1
        return num1 - num2

    def Mul (self, num1, num2):
        self.MulCount += 1
        return num1 * num2

    def Div (self, num1, num2):
        self.DivCount += 1
        return num1 / num2

    def ShowCount(self):

        print("덧셈 : %s" %self.AddCount)
        print("뺄셈 : %s" %self.SubCount)
        print("곱셈 : %s" %self.MulCount)
        print("나눗셈 : %s" %self.DivCount)

c = Calculator()

print(c.Mul(100,5))

print(c.ShowCount())
```

## Private variables, Getter, Setter
1. Private Variable: class에서만 쓰이는 변수. 내부의 변수를 외부에서 access 못할때 '__변수이름' 사용
- 보안을 위해 사용된다.
- 내부의 변수를 외부로 함부로 가져가지 못하도록 한다.
- Getter를 통해서만 접속이 가능하다.
2. Getter : private variable 값을 가져오는 함수
3. Setter : private variable 값을 변경해주는 함수

```python
class Circle :


    def __init__(self, r):
        self.__radius = r # private variable을 설정해 준다.

        # class 안에 변수 만들기.
        Circle.pi = 3.14

    def area(self):
        return Circle.pi * self.__radius**2

    def circumference(self):
        return Circle.pi * self.__radius * 2

    
    @property
    def radius(self):
        return self.__radius
    
    @radius.setter
    def radius(self, value):
        if value <= 0:
            raise TypeError("양의 숫자만 입력하세요")
        self.__radius = value
        
       
        
circle = Circle(10)
print("반지름:",circle.radius ) # __radius가 프라이빗 변수임에도 불구하고 값을 출력한다
circle .radius = 2
print("setter를 이용해서 변경한 반지름", circle.radius)
circle.circumference()
circle.radius = -2
```

## 상속
- 상속은 이미 존재하는 클래스에서 필요한 기능(함수)만 가져오는 것을 의미한다.
- 클래스 종류:
1. Parent(부모) 클래스: 기능을 가져오는 클래스 = Super class
2. Child(자식) 클래스: 기능을 이식하는 (대상) 클래스 = Sub class

- Child 클래스 안에 'super().매소드_이름()'를 입력하면 동일한 이름의 Parent 클래스도 호출할 수 있다.
- 클래스는 다중상속 가능!: 여러 Parent 클래스에서 여러 기능들을 가져오는 것 -> C언어나 Java는 불가능

- Override: 상속 받아 온 매소드를 자식이 '재정의'하는 것

```python
# Parent Class
class Country:
    "parent"
    name = "국가명"
    population = "인구"
    ccapital = "수도"
    
    def show(self):
        print("국가 클래스 매소드")

# Child Class
class Korea(Country):
    def __init__(self,name):
        self.name = name
    def show_name(self):
        print("국가 이름", self.name)
        
a = Korea("대한민국") # Child 클래스인 Korea를 불러옴
a.show() # 그러나 상위 클래스인 show()에서 불러옴 -> a를 child 클래스로 할당했음에도, parent 클래스 매소드인 show()를 불러왔음
a.show_name()
```

```python
# Overriding

# Parent Class
class Country:
    "parent"
    name = "국가명"
    population = "인구"
    ccapital = "수도"
    
    def show(self):
        print("국가 클래스 매소드")

# Child Class
class Korea(Country):
    def __init__(self,name,population,capital):
        self.name = name
        self.population = population
        self.capital = capital
    def show(self): # Parent 클래스의 show() 매서드와 같은 이름을 가지고 있음
        print(f"""
        이름: {self.name}
        인구: {self.population}
        수도: {self.capital}
        """)
# 동일한 이름의 매소드가 있는 경우 하위 Child 클래스의 매소드가 우선적으로 산출된다.
a = Korea("대한민국", 50000000, "서울") # Child 클래스인 Korea를 불러옴
a.show() # Parent 클래스를 가져왔음에도 Child 클래스에서 동일한 매소드 이름으로 갱신을 하면 Overrinding되어 새로운 값을 가져온다.
```

```python
# super(): 이를 이용하면, 동일한 이름의 Parent 클래스를 호출할 수 있다.
# Child 클래스 안에 super().매소드_이름을 코드로 입력하면 Parent 클래스를 불러올 수 있다.

# Parent Class
class Country:
    "parent"
    name = "국가명"
    population = "인구"
    ccapital = "수도"
    
    def show(self):
        print("국가 클래스 매소드")

# Child Class
class Korea(Country):
    def __init__(self,name,population,capital):
        self.name = name
        self.population = population
        self.capital = capital
    def show(self): # Parent 클래스의 show() 매서드와 같은 이름을 가지고 있음
        super().show() # Parent 클래스 호출 가능
        print(f"""
        이름: {self.name}
        인구: {self.population}
        수도: {self.capital}
        """)
a = Korea("대한민국", 50000000, "서울")
a.show()        
```
