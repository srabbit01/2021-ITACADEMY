## 여백(Margin) 스타일
1. margin 속성으로 요소 주변 여백 설정하기

- 4개의 면을 각각 지정할 수 있다: 시계 방향(top/right/bottom/left) -> 테두리와 적용이 같다.

2. margin 속성으로 웹 문서 가운데 정렬하기
- 문서는 기본적으로 왼쪽에 정렬된다.
- 설정 방법  
① 배치 요소의 너비(width) 값이 정해져 있어야 한다.  
② 마진 옵션을 정해준다: margin-left | margin-right | auto  
- Auto: 화면너비 - 요소너비 = 좌우마진 (중앙 배치)
3. margin 중첩
- 요소들을 세로(위/아래)로 배치하면 마진 값이 큰쪽이 작은쪽을 덮는다.
- 예를 들어, 위 마진이 30px 아래 마진이 60px라 할 때 사이 간격은 60px가 된다.
- 그러나, 가로(좌/우) 배치에는 중첩이 일어나지 않는다.


```python
# margin
margin: 크기(px,em)|%|auto
```

## 웹문서 레이아웃 만들기
1. Display(배치 방법): 블록 레벨 요소/인라인 레벨 요소를 바꿔서 쓸 수 있다.
- 역할:  
① 네비게이션 메뉴: 메뉴 항목을 가로로 배치할 때  
② 이미지를 표 형태로 배치할 때  
- Option(옵션): block | inline | inline-block | none  
① block: 블록 레벨로 설정  
② inline: 인라인 레벨로 설정  
③ inline-block: 블록/인라인 요소 설정  
④ none: 아무것도 표시하지 않음  
2. Float: 왼쪽/오른쪽 표시
- Option(옵션): left | right | none(기본값)
- Clear: left | right | both(양쪽 모두 해제)
- float을 깨기 위해서는 clear를 사용 -> float을 한번 선언할 경우, 다음 요소에도 같은 속성이 전달되기에 필요가 없어지면 clear를 사용하여 깨준다.  
예시) 이미지를  paragraph <p'> 태그를 쓰는거 처럼 나란히 왼쪽/오른쪽에 정렬할때 -> 이미지 옆에 글자 넣기 가능
3. 레이아웃:  
① 레이아웃을 짠다. (전체가 container)  
  
② HTML / CSS 나란히 놓기  
- float을 header 아래 body에서 지정해 주면, footer에서 clear를 이용하여 깨주어야 한다.
- 주의사항: 내용 전체 너비 기준으로 내부 값을 잘 지정해야 한다.

## 웹요소 위치 지정하기
- 외부의 요소를 이미지 등 내부 요소에 넣을 수 있다.
1. postion 속성: margin 외부 속성

- static: 기본값. 문서 흐름에 맞추어 배치
- relative : static + 위치값을 정해줄 수 있다.

- absolute : relative 값을 사용한 상위요소를 기준으로 위치를 지정 (고정값) -> relative 지정하지 X시 브라우저 기준
- fixed : 브라우저 창을 기준으로 위치 지정 (고정값)
2. left/right/top/bottom 속성
- 반드시 기준 위치(postition)가 있어야 한다. 
- left: left 옵션을 준 요소가 '기준 위치'로부터 왼쪽으로 얼마나 떨어져 있는가를 지정
- right: right 옵션을 준 요소가 '기준 위치'로부터 오른쪽으로 얼마나 떨어져 있는가를 지정
- top: top 옵션을 준 요소가 '기준 위치'로부터 얼마나 위쪽으로 떨어져 있는가를 지정
- bottom: bottom 옵션을 준 요소가 '기준 위치'로부터 아래쪽으로 얼마나 떨어져 있는가를 지정  
ex)

- 둘 이상의 선택자를 연결하여 스타일이 적용될 요소가 어느 부분인지 결정한다.

- 공백으로 상위요소와 하위요소를 구분한다.
  

② 자식 선택자: 오직 자식(바로 하위 요소)에게만 스타일 적용

③ 형제(sibling) 선택자: 모든 형제 요소 선택

ex) h1 ~ p{}  

④ 인접 형제(adjacent): 형제 중에서 '첫번째' 동생에게만 스타일 적용

ex) 
2. 속성 선택자  

① 특정 속성 선택  

ex) <a'> 태그 중에서 href 속성을 가지고 있는 태그 선택:

② 특정 속성값 선택: 특정 속성이 오직 특정 속성값만 가져야 적용된다.

ex) A[target = _blank] {     }  

③ 여러 값 중에서 '특정 속성값이 포함된 속성' 요소 선택: 특정 속성값 안에 특정 속성값이 포함되기만 하면 적용된다.

ex) 
- 만일 '='이었으면 노란색만 선택되지만, '~='이면 노란색+초록색 모두 선택된다.  

④ 특정 속성값이 포함된 속성요소 선택: 특정 속성값을 정확히 모두 입력하지 않아도 선택된다.

ex) 
⑤ 특정 속성값으로 시작하는 속성 요소 선택

ex) 
⑥ 특정 속성값으로 끝나는 속성 요소 선택 -> 보통 확장자 기준 선택할 때

⑦ 특정 속성값 중 일부가 일치하는 속성 요소 선택

ex) 

```python
# 하위 선택자
상위요소 하위요소{}

# 자식 선택자
상위요소 > 하위요소{}

# 형제 선택자
요소1~요소2{}
ex) h1 ~ p{}

# 인접 형제
요소1+요소2{}
ex) h1 + p{}

# 특정 속상 선택
태그 [찾고싶은_속성] {}

# 특정 속성값 선택
태그 [찾고싶은_속성 = 찾고_싶은_속성이_가지고_있는_속성값] {}

# 여러 값 중에서 특정 속성값이 포함된 속성 요소 선택
태그 [찾고싶은_속성 ~= 찾고_싶은_속성이_가지고_있는_속성값] {}

# 특정 속성값이 포함된 속성요소 선택
태그 [찾고싶은_속성 |= 찾고_싶은_속성이_가지고_있는_속성값] {}

# 특정 속성값으로 시작하는 속성 요소 선택
태그 [찾고싶은_속성 ^= 찾고_싶은_속성이_가지고_있는_속성값] {}

# 특정 속성값으로 끝나는 속성 요소 선택
태그 [찾고싶은_속성 $= 찾고_싶은_속성이_가지고_있는_속성값] {}

# 특정 속성값 중 일부가 일치하는 속성 요소 선택
태그 [찾고싶은_속성 *= 찾고_싶은_속성이_가지고_있는_속성값] {}
```
