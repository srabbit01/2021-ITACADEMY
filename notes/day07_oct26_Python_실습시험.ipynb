## 문제 1. 행맨 게임을 코딩하시오
- 조건 1. 행맨의 전체 작업 플로우를 Diagram으로 그리고 코드와 함께 제출한다
- 조건 2. 사용자가 이미 입력한 글자는 다시 입력할 수 없도록 제한한다


```python
from string import ascii_lowercase
alpha_list = list(ascii_lowercase)

# 단어 랜덤 추출
fruit = ["buleberry", "orange", "melon", "lemon", "peach"]
import random
ran_fruit = random.choice(fruit)
fruit_list = ('_ '*len(ran_fruit)).split( )
# 도입부
print(fruit_list, "Life: 5")
print("주의: 같은 알파벳을 반복적으로 입력할 시, Error가 발생합니다.")
# 행맨 만들기
life = 5
fruit_len1 = len(ran_fruit)
fruit_len2 = len(ran_fruit)
while not fruit_len1 == 0:
    print()
    alpha = input("알파벳을 입력하세요: ")
    if alpha not in alpha_list:
        print("Error")
        break
    else:
        pass
    for i in range(0,len(ran_fruit)):
        if alpha == ran_fruit[i]:
            fruit_list[i] = alpha
            fruit_len1 -= 1
    if fruit_len1 == fruit_len2:
        print("Life가 하나 차감되었습니다.")
        life -= 1
    fruit_len2 = fruit_len1
    print(fruit_list, "Life:",life)
    alpha_list.remove(alpha)
print()
if life == 0:
    print("Death")
if fruit_len1 == 0:
    print("Congratulations! 단어 %s를 완성하셨습니다." %ran_fruit)
```

```python
code = {'hello_python': '3!2F2d', 'id_is_so_hard': '&f)O4&0)h', 'hello_computer': 'L)5RpcQ*'}

def hangman():
    from string import ascii_lowercase
    alpha_list = list(ascii_lowercase)
    fruit = ["blueberry", "orange", "melon", "lemon", "peach"]
    import random
    ran_fruit = random.choice(fruit)
    fruit_list = ('_ '*len(ran_fruit)).split( )
    print(fruit_list, "Life: 5")
    print("주의: 같은 알파벳을 반복적으로 입력할 시, Error가 발생합니다.")
    life = 5
    fruit_len1 = len(ran_fruit)
    fruit_len2 = len(ran_fruit)
    while not fruit_len1 == 0:
        print()
        alpha = input("알파벳을 입력하세요: ")
        if alpha not in alpha_list:
            print("Error")
            break
        else:
            pass
        for i in range(0,len(ran_fruit)):
            if alpha == ran_fruit[i]:
                fruit_list[i] = alpha
                fruit_len1 -= 1
        if fruit_len1 == fruit_len2:
            print("Life가 하나 차감되었습니다.")
            life -= 1
        fruit_len2 = fruit_len1
        print(fruit_list, "Life:",life)
        alpha_list.remove(alpha)
    print()
    if life == 0:
        print("Death")
    if fruit_len1 == 0:
        print("Congratulations! 단어 %s를 완성하셨습니다." %ran_fruit)
        
game = False
no = 3
while not no == 0 and game == False: 
    print(f"입력 기회는 {no}번 남았습니다.")
    ID = input("아이디를 입력하세요: ")
    Password = input("비밀번호를 입력하세요: ")
    for i in code:
        if i == ID and code[i] == Password:
            print()
            print("행맨을 시작합니다.")
            hangman()
            game = True
            no -= 3
            break
    if game:
        break
    print("아이디 혹은 비밀번호를 다시 입력하세요.")
    no -= 1
    print()
```
