# 4. 선형 모형
- "선을 그어서 데이터 예측하기"
- 가장 기본적인 분석 중 하나로 선형 회기 분석을 꼽을 수 있다.
1. 선형 회기식
- y 값을 예측하기 위해서는 '특성 x, x가 가지는 가중치, 상수' 세가지를 고려해야 한다.
- 함수: 두 집합 사이의 관계를 설명하는 수학 개념

- y = 종속변수(예측값) / x = 독립변수(입력값) / a = 계수(x 기울기: 가중치) / b = 상수
- 즉, x와 y의 관계를 표현한 것이다.
2. 선형회기 기초
- 선형 모델은 던체 데이터셋에서의 "중심선"을 찾는 것을 목적으로 한다.
- 중심선: 선의 형태를 갖기 때문에 '선형' 모델이라 부른다.

- 선형 모델의 중심선을 함수로 표현하면 
- 하나 이상의 특성(x)를 가지게 되는 경우를 회기선으로 표현하면 

3. MSE(Mean Squared Error) - 회기선 긋는 방법  
① 참값과 예측값의 차이(Error)를 모두 구한다.   

② 모든 에러를 각자 제곱(Squared,자승)해 준다.


③ 제곱한 에러를 모두 합해 평균(Mean)을 구한다.  
④ 제곱한 에러의 평균 값이 제일 작아지도록 (최소) 회귀선을 그린다.

4. ** 계수와 상수  
① 계수(Codefficient): 회귀선을 그릴 때 회귀선의 기울기(가중치) -> 가중치는 +/-의 방향성과 그 크기에 대한 정보를 둘 다 담고 있기 때문에 가장 중요한 정보이다.  
② 상수(Constant): 시작점의 높낮이  

5. 경사 하강법: 최적의 가중치 구하기
- MSE(오차 평균)가 가장 작은 값을 찾기 위해 그리는 회귀선
- 경사 하강법은 '기울기사 감소하는 방향'으로 값을 조정하는 방법이다

- 가중치(a)가 너무 커도, 너무 작아도 오차가 커지기 때문에 적정한 가중치가 필요 -> 경사 하상법 사용

① 가중치를 미분하여 해당 가중치의 기울기가 0인 것을 찾아 조절한다.

② 곡선 중에서 오차가 가장 작은 부분을 찾아 구해준다.

6. 가중치 규제
- 선형 회기 모형은 과적한 발생이 쉽기 때문에 1) 특성 변수의 수를 줄이거나 2) 모델의 복잡도를 강제로 줄여야 한다.
- 이 중 2) 모델의 복잡도를 줄이기 위해 각 특성 변수의 가중치를 줄여야 한다.

- L1, L2 규제의 크기는 'alpha' 파라미터로 정해진다.
- 가중치 규제법 종류:  
① L1 규제  
  - 가중치의 합을 더한 값에 규제 강도 λ를 곱한 값을 오차에 더한다.
  - λ의 크기에 비례해 규제가 커진다.
  - 어떤 가중치는 0이 될 수도 있다.
  - 가중치가 적은 특성(중요하지 않은 특성)을 모델에서 완전히 제외할 수도 있다는 의미 -> 특성 변수가 많은데 중요하지 않은 변수들이 있다고 의심될 때, 혹은 분석시에 어떤 독립변수가 중요한지 모를 때 적용한다.  

  ② L2 규제  
  - 각 가중치 제곱의 합에 규제 강도 λ를 곱한다.
  - λ값이 커지면 규제도 커지고, 규제가 커지면 가중치가 줄어들어 모델이 단순해 진다.
  - 가중치 제곱의 합에 규제가 곱해지도록 가중치가 클 수록 규제도 상대적으로 크게 적용한다.
  - "노이즈나 이상치가 많은 데이터셋을 분석할 때" 강점을 보인다.
 
- 가중치 규제외 회귀 모형:  

① L1 규제와 라소 회귀(Lasso Regression)
  - 라소 회귀: 가중치에서 상수 값을 제거하는 방식 -> 큰 역할을 하지 않는 가중치는 0이다.
  - 모델이 알아서 중요 독립변수만 걸러 주기 때문에 독립변수가 많지만 일부만 중요 변수일 경우 좋은 성능을 낼 수 있다.
  - 모델의 복잡도에 영향을 크게 받는 편이다.
  - Alpha 값을 줄여서 모델을 복잡하게 만들면 성능이 좋아지는 경우가 많다.  
  
  ② L2 규제와 리지 회귀(Ridge Regression)  
  - 가중치를 작게 만들어 과대적합을 피한다.
  - 특성상 큰 가중치에는 큰 규제를, 작은 가중치를 작은 규제를 가한다.
  - 특이치/이상치에 강하다.
  
- 최적의 alpha값 설정하기
  - 규제강도 = 'alpha값'
  - 라소/리지 모형 사용시 반드시 alpha 값을 튜닝해 주어야 한다.
  
- 훈련 데이터의 크기
  - 회귀 모델들은 훈련 데이터 크기의 영향을 많이 받는다. -> 데이터 크기가 클 수록 성능이 좋음
  - 선형 회귀는 훈련 데이터가 일정량 이상 (특성당 최소 10개 이상) 되어야 학습할 수 있다. -> 샘플이 많을 수록 빠르게 성능이 좋아짐
  - 리지 회귀는 훈련 데이터가 적어도 비교적 성능이 빨리 좋아지는 편이다.
  - 데이터가 충분히 많으면 일반 선형 회귀 모형이 가장 일반적인 적합도를 지니면서 성능이 좋다. (일반적으로 특성당 200개 이상)
  - 훈련 데이터가 많아지면 성능은 일반 선형 회귀 모형 > 리지/라소 모형이다.
  - 훈련 데이터가 충분하면 독립 변수가 많아도 과대적합의 가능성이 줄어든다.

```python
# 가상의 데이터셋 만들기
import matplotlib.pyplot as plt
from sklearn.datasets import make_regression

# 회귀용 데이터셋을 생성한다
# 샘플=100개, 특성=1개, 편향도=10, 노이즈=10
x, y, c = make_regression(n_samples=100, n_features=1, # n_sample='추출할_샘플수' # n_features='독립변수_갯수'
                          bias=100, noise=20, coef=True, # bias: y절편  # noise: 출력 정규분포의 표준차 # coef: Trueaus 선형 모형의 계수도 출력
                          random_state=0) # random_state: 난수 발생용 시작값
plt.scatter(x, y, s=50, c='orange') # plt.scatter(): 산점도 # s(markersize) = '점의 사이즈' # c(color) = '점의 색깔'
plt.show()

print("c\n", c) # c(coefficient): 선형 모형의 계수(기울기)
```

> make_regression(): 가상의 회기 분석 문제 만들기  
    - bias: y 절편  
    - noise: 출력 즉, 종속 변수에 더해지는 정규 분포의 표준 편차  
    - coef: 선형 모형의 계수(기울기). True면 선형 모형의 계수도 출력  
    - random_state: 난수 발생용 시작값
    

> plt.scatter(): 산점도 그리기  
     - s(markersize) = '점의 사이즈'  
     - c(color) = '점의 색깔'

```python
# Step 3. 데이터를 훈련셋과 테스트셋으로 분리한다

# 훈련데이터와 테스트 데이터의 분리
from sklearn.model_selection import train_test_split
X_train, X_test, y_train, y_test = train_test_split(x,y,random_state=0)
```

#### 4-1. 선형 회귀 모델

```python
# 분석을 위해 선형회귀 모델을 불러온다
from sklearn.linear_model import LinearRegression # LinearRegression: 선형 회귀 모델

# 호출한 선형회귀 분석을 훈련 데이터셋으로 훈련하고 테스트 데이터로 예측한다
lr = LinearRegression().fit(X_train, y_train)

y_pred = LinearRegression().fit(X_train, y_train).predict(X_test) # 테스트 데이터 예측


# Step 4. 모델을 검정한다
# r2 스코어를 내기 위한 모듈 불러오기
from sklearn.metrics import r2_score # r2_score: 결정계수. 1에 가까울 수록 상관관계가 높음을 의미
# r2 스코어 출력하기
print("테스트 점수(r2) :", round(r2_score(y_pred,y_test),3))
print("기울기", lr.coef_) # coef_: 기울기
print("절편", lr.intercept_) # intercept_: y 절편
```

#### 4-2. 라소 모델

```python
# 분석을 위해 라쏘 모델을 불러온다
from sklearn.linear_model import Lasso # Lasso: 라소 모델

# 호출한 선형회귀 분석을 훈련 데이터셋으로 훈련하고 테스트 데이터로 예측한다
lr = LinearRegression().fit(X_train, y_train)
y_pred = Lasso().fit(X_train, y_train).predict(X_test)

# Step 4. 모델을 검정한다
# r2 스코어를 내기 위한 모듈 불러오기
from sklearn.metrics import r2_score
# r2 스코어 출력하기
print("테스트 점수(r2) :", round(r2_score(y_pred,y_test),3))

# 라쏘모형은 기울기 값이 0임
```

#### 4-3. 릿지 모델

```python
# 분석을 위해 릿지 모델을 불러온다
from sklearn.linear_model import Ridge

# 호출한 선형회귀 분석을 훈련 데이터셋으로 훈련하고 테스트 데이터로 예측한다
y_pred = Ridge().fit(X_train, y_train).predict(X_test)


# Step 4. 모델을 검정한다
# r2 스코어를 내기 위한 모듈 불러오기
from sklearn.metrics import r2_score
# r2 스코어 출력하기
print("테스트 점수(r2) :", round(r2_score(y_pred,y_test),3))

# 릿지모형은 기울기 값이 0으로 수렴함
```

#### 4-4. 최적의 alpha값 찾기
- Test Score이 가장 높은 것이 최적의 alpha값이다.


```python
# 라소 모형

import numpy as np
from sklearn.linear_model import Lasso
import matplotlib.pyplot as plt

# 결과 점수를 저장할 리스트를 선언한다
train_accuracy = []
test_accuracy = []

# 알파의 값을 0에서 10까지, 0.1씩 증가하면서 만들어 준다
alphas = np.arange(start=0, stop=10, step=0.1) # 0에서 시작해서 10전까지 0.1씩 증가하는 alpha값들의 리스트를 만들어 준다

# For loop 을 이용하여 앞서 생성한 알파값을 하나씩 적용해본다
# 이때 하나씩 적용한 알파값의 결과는 각자 선언한 리스트에 저장한다
for alpha in alphas:
    lasso = Lasso(alpha=alpha).fit(X_train, y_train) # alpha값들을 차례로 넣어서 훈련해 준다
    train_accuracy.append(lasso.score(X_train, y_train)) # 훈련 결과를 담는다
    test_accuracy.append(lasso.score(X_test, y_test)) # 테스트 결과를 담는다

# 분석결과를 시각화 해 본다    
plt.plot(alphas, train_accuracy, label="Training Score")
plt.plot(alphas, test_accuracy, label="Test Score")
plt.legend() # 범례 표기
plt.show()
```

```python
# 리지 모형

import numpy as np
from sklearn.linear_model import Ridge
import matplotlib.pyplot as plt

# 결과 점수를 저장할 리스트를 선언한다
train_accuracy = []
test_accuracy = []

# 알파의 값을 0에서 30까지, 0.1씩 증가하면서 만들어 준다
alphas = np.arange(start=0, stop=30, step=0.1)

# For loop 을 이용하여 앞서 생성한 알파값을 하나씩 적용해본다
# 이때 하나씩 적용한 알파값의 결과는 각자 선언한 리스트에 저장한다
for alpha in alphas:
    ridge = Ridge(alpha=alpha, max_iter=2000).fit(X_train, y_train) # max_iter: 반복 실행하는 최대 횟수
    train_accuracy.append(ridge.score(X_train, y_train))
    test_accuracy.append(ridge.score(X_test, y_test))

# 분석결과를 시각화 해 본다    
plt.plot(alphas, train_accuracy, label="Training Score")
plt.plot(alphas, test_accuracy, label="Test Score")
plt.legend()
plt.show()
```
