```python
#   
import matplotlib.pyplot as plt
from sklearn.datasets import make_regression

#   
# =100, =1, =10, =10
x, y, c = make_regression(n_samples=100, n_features=1, # n_sample='_' # n_features='_'
                          bias=100, noise=20, coef=True, # bias: y  # noise:    # coef: Trueaus    
                          random_state=0) # random_state:   
plt.scatter(x, y, s=50, c='orange') # plt.scatter():  # s(markersize) = ' ' # c(color) = ' '
plt.show()

print("c\n", c) # c(coefficient):   ()
```

> make_regression():       
    - bias: y   
    - noise:  ,         
    - coef:   (). True      
    - random_state:   
    

> plt.scatter():    
     - s(markersize) = ' '  
     - c(color) = ' '

```python
# Step 3.    

#    
from sklearn.model_selection import train_test_split
X_train, X_test, y_train, y_test = train_test_split(x,y,random_state=0)
```

#### 4-1.   

```python
#     
from sklearn.linear_model import LinearRegression # LinearRegression:   

#         
lr = LinearRegression().fit(X_train, y_train)

y_pred = LinearRegression().fit(X_train, y_train).predict(X_test) #   


# Step 4.  
# r2     
from sklearn.metrics import r2_score # r2_score: . 1     
# r2  
print(" (r2) :", round(r2_score(y_pred,y_test),3))
print("", lr.coef_) # coef_: 
print("", lr.intercept_) # intercept_: y 
```

#### 4-2.  

```python
#     
from sklearn.linear_model import Lasso # Lasso:  

#         
lr = LinearRegression().fit(X_train, y_train)
y_pred = Lasso().fit(X_train, y_train).predict(X_test)

# Step 4.  
# r2     
from sklearn.metrics import r2_score
# r2  
print(" (r2) :", round(r2_score(y_pred,y_test),3))

#    0
```

#### 4-3.  

```python
#     
from sklearn.linear_model import Ridge

#         
y_pred = Ridge().fit(X_train, y_train).predict(X_test)


# Step 4.  
# r2     
from sklearn.metrics import r2_score
# r2  
print(" (r2) :", round(r2_score(y_pred,y_test),3))

#    0 
```

#### 4-4.  alpha 
- Test Score     alpha.


```python
#  

import numpy as np
from sklearn.linear_model import Lasso
import matplotlib.pyplot as plt

#     
train_accuracy = []
test_accuracy = []

#   0 10, 0.1   
alphas = np.arange(start=0, stop=10, step=0.1) # 0  10 0.1  alpha   

# For loop       
#         
for alpha in alphas:
    lasso = Lasso(alpha=alpha).fit(X_train, y_train) # alpha    
    train_accuracy.append(lasso.score(X_train, y_train)) #   
    test_accuracy.append(lasso.score(X_test, y_test)) #   

#        
plt.plot(alphas, train_accuracy, label="Training Score")
plt.plot(alphas, test_accuracy, label="Test Score")
plt.legend() #  
plt.show()
```

```python
#  

import numpy as np
from sklearn.linear_model import Ridge
import matplotlib.pyplot as plt

#     
train_accuracy = []
test_accuracy = []

#   0 30, 0.1   
alphas = np.arange(start=0, stop=30, step=0.1)

# For loop       
#         
for alpha in alphas:
    ridge = Ridge(alpha=alpha, max_iter=2000).fit(X_train, y_train) # max_iter:    
    train_accuracy.append(ridge.score(X_train, y_train))
    test_accuracy.append(ridge.score(X_test, y_test))

#        
plt.plot(alphas, train_accuracy, label="Training Score")
plt.plot(alphas, test_accuracy, label="Test Score")
plt.legend()
plt.show()
```
