# 가상 클래스 & 가상 요소
1. '사용자 동작'에 반응하는 가상 클래스: “LoVe HAte”
- :link: 방문하지 않은 링크에 적용 (파란색)
- :visited: 방문한 링크에 적용 (빨간색)
- :hover: 특정 요소에 마우스 포인터를 올려 놓으면 적용 (해당 부분의 색상 등이 변함)
- :active: 웹 요소를 활성화 했을 때 = 클릭했을 때
- :focus: 웹 요소에 초점이 맞추어졌을 때 = 클릭하고 땠을 때 (해당 부분이 눌린 상태로 있을 때)

2. '요소 상태'에 따른 가상 클래스
- :target: 앵커 대상에 스타일 적용
- :enabled: 요소가 있는데 사용할 수 있을 때 (요소 활성화)
- :disabled: 요소가 있는데 사용하지 못할 때 (요소 비활성화)
- :checked:  checkbox/ radio button 옵션이 활성화 되었을 때 적용. 즉, check 박스에 check를 했을 때 새로운 스타일을 적용하는 것  
ex) check 사 해당 글자 색상이 바뀌는 것 등
- :not: 특정 요소를 제외할 때 적용 
3. '구조' 가상 클래스

- 'A'는 요소를 의미한다.
4. 가상 요소 (콜론 두 개 :: 로 표시)
- 문서 안에 특정 부분을 지정하는 것 -> 불필요한 태그(span 등) 사용을 줄일 수 있다.
- ::first-line/::first-letter: 내용 중에서 첫번째 줄/글자에 적용 -> note: 반드시 첫번째 줄에 내용이 있어야 한다.  
ex) <p'> 내용1 <br'> 내용2 </p'> -> 내용2에서는 적용이 되지 않는다.
- ::before/::after: 내용 앞뒤에 컨텐츠를 추가

```python
# 응용 예시
.topmenu li:not(:last-child) a: hover{}
```
