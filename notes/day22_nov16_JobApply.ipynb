## 입사 지원서 작성  
: 입사지원서는 수많은 지원자 중 적합한 인재를 선별하고 면접의 기초자료를 활용하기 위한 자료이다.
- 사진을 보지 않는 블라인드 채용을 하나, 아직도 사진 첨부가 중요한 요소이다. (최근 3개월 이내)
- 사진은 '파란색 배경'으로 하는 것이 좋다.
- 교육/직무능력/경력사항: 가장 최근 것부터 작성한다.
- 직무능력의 경우, 무엇을 활용할 수 있는지 매우 자세하게 작성한다.
- 나이에 따라서, (어린 경우) 3개월 이상 한 일(알바 등)은 적기도 한다.
- 주의사항  
① 간결하고 일관성 있게 작성  
② 과장이나 거짓말 금지 -> 장점은 부각하되, 과장은 좋지 않음  
③ 오타 및 문법 반드시 확인  
④ 내용뿐만 아니라 시각적 매력도 느껴지도록 깔끔하게 정리  
⑤ 자격증 취즉년도, 교육이수 과목명 등 정확히 명시  
- 직무와 알맞게 구체적으로 작성해야 한다.
- 자기만의 스토리로 자기소개서를 작성해야 한다.
- 자기소개서에서 가장 중요한 것은 지원동기이다.
- 자기소개서는 쓸 때 반드시 핵심 키워드를 정하고 작성한다.

1. 지원동기 **
- 회사와 직무에 대해 제대로 이해하고 있는가?
- 입사하고자 하는 확고한 의지가 있는가?
- 지원자의 직무역량, 직무적합성 파악
- 계기 -> 어색 -> 흥미 느낌 -> 더 공부하고자 마음 먹음 -> 능록(경험) -> 배움을 거듭 -> 즐거움 -> 꿈 확고 -> 앞으로 회사에서 어떻게 할 것인가
2. 입사 후 포부
- 직장 생활에서 이루고자 하는 구체적인 목표는 무엇인가?
- 주어진 업무는 어떤 아이디어나 각오로 해낼 것인가?
- 000 회사 000 직무에서 다음과 같이 행하겠습니다. 첫째 ~ 둘째 ~
3. 성격 및 장단점
- 본인의 장단점을 정확하게 파악하는가?
- 장점의 경우 여러개 나열하지 말고, 구체적으로 경험과 함께 적기
- 단점이 있다면 그 상황에 대한 대비책을 보유하고 있는가? -> 사람이 가진 일반적인 단점 적기
4. 성장과정
- 지원자가 직무에 적합한 성향과 가치관을 지니고 있는가?
- 회사의 구성원이 되었을 때 미치는 영향 파악
- 경험을 바탕으로 어필하고 싶은 이미지를 하나의 사건을 중심으로 작성
