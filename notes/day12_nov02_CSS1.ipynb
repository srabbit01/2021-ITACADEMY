## 스타일 종류

1. color: 글자 색상
2. background-color: 배경 색상
3. padding: 요소 내부의 빈 공간

 - 한 개의 값은 모든 네 면의 여백을 설정합니다.
 - 두 개의 값을 지정하면 첫 번째는 위와 아래, 두 번째는 왼쪽과 오른쪽 여백을 설정합니다.
 - 세 개의 값을 지정하면 첫 번째는 위, 두 번째는 왼쪽과 오른쪽, 세 번째 값은 아래 여백을 설정합니다.
 - 네 개의 값을 지정하면 각각 상, 우, 하, 좌 순서로 여백을 지정합니다. (시계방향)
4. margin:요소 주위의 빈 공간
5. width: 가로 넓이/ height: 세로 넓이  
   ① px: 픽셀 단위로 절대적인 크기 설정 (브라우저 크기에 따라 변하지 않음)  
   ② %: 브라우저 너비/높이를 기준으로 % 크기 설정  
6. border(테두리): 크기 종류 색상  
ex) border:1px solid #eee


7. border-width: 테두리 넓이
8. border-radius: 테두리 모양


## CSS 기본 선택자
1. 전체 선택자: 문서에서 해당 태그 값을 가지는 모든 요소(element)에 적용 -> '*'(별표)를 이용하여 표시해 준다.

2. 타입(type) 선택자: 특정 '**'태그를 사용한 모든 요소'*'에 스타일을 적용한다.

3. 클래스 선택자

- 지정한 클래스에 적용되는 것들은 스타일을 적용한다.
- 요소 하나당 하나 이상의 클래스를 지정해 줄 수 있다. -> 여러가지의 스타일을 동시에 적용 가능

4. ID 연산자

- 특정 ID를 지정할 수 있으며, 문서에서 한번만 적용 가능하다.
- 클래스와의 차이: 클래스는 여러번 반복적으로 지정이 가능하나, ID는 한번만 지정이 가능하다.

5. 그룹 선택자: 하나 이상의 선택자에 동일한 스타일 규칙 적용 할 때 사용한다.



```python
# 전체 선택자
*{속성1:속성값1;속성2:속성값2;...}

# 타입 선택자
type{속성1:속성값1;속성2:속성값2;...}
p{
    font-style: italic;
    font-size: larger;
}

# 클래스 선택자
.class명{스타일_규칙}

# ID 연산자
#ID{속성1:속성값1;속성2:속성값2;...}

# 그룹 선택자
선택자1,선택자2{속성1:속성값1;속성2:속성값2;...}
```

## 캐스캐이딩 스타일 시트 ***
: 한 요소에 적용되는 스타일이 둘 이상일때, 우선순위가 계단식으로 적용된다.
1. 스타일 우선순위: ① 스타일 중요도 ② 적용 범위에 따라서 높은 계급부터 우선 적용  
   ① 스타일 중요도: 사용자 > 제작자 > 브라우저
   
   ② 적용 범위: !important > 인라인 스타일 > id 스타일 > class 스타일 > type 스타일
   
   - 중요도와 적용 범위가 모두 같을 경우: 작성한 코드 순서대로 우선순위가 적용된다. ('나중'에 쓴 것이 먼저 쓴 것을 덮어씀)
   - '!important'는 작성한 것 오른쪽에 넣으면 된다.
2. 스타일 상속: 태그의 포함관계에 따라 부모 요소의 스타일을 자식 요소에게 전달  
- HTML에는 포함 관계가 있으며 상위 요소는 부모, 하위 요소는 자식이라 한다.
- 이때, 부모의 스타일 요소는 자식에게 전달이 되며 이를 스타일 상속이라 한다.
- 단, 자식에게 지정된 스타일이 있다면 그것이 우선시 된다. -> 부모 !important < 자식
- NOTE : 단, 배경색/배경이미지는 상속이 안된다.

## 글꼴(font) 관련 스타일
1. 글꼴 지정: font-family


2. 글자 크기 설정: font-size

 ① 절대 크기: 브라우저에서 지정한 크기  
 ② 상대 크기: 부모의 글자 크기를 기준으로 상대적인 글자 크기 지정 (최근에는 많이 사용)  
 ③ 크기: 브라우저하고 상관없이 글자 크기 지정  
 ④ 백분율: 부모의 글자 크기를 기준으로 설정한 백분율(%)  
- font-size 예약어들: ex) font-size: small -> 정해진 글꼴 기준


- font-size 지정 단위: 최근에는 해상도 등 문제로 인해 상재적인 크기(em/rem)을 많이 사용한다.

- NOTE: 백분율(%)을 사용하려면 부모의 글꼴 크기가 pt/px 단위로 표현되어 있어야 한다.
- 맨 처음의 기준은 루트 값을 이용하여 설정한다. (브라우저의 완전 기본 글자 크기: 12pt = 16px)
3. 글자 이탤릭체 적용

- normal
- italic: 설정된 font를 기울인 것이다.
- oblique: 현재 사용하고 있는 font를 기울인다.
4. 글자 굵기 설정: font-weight

- bolder: 더 굵게
- light: 가늘게
- lighter: normal을 기준으로 더 가늘게
- 굵기 수치: 100(제일 얇음) ~ 900(제일 굵음)
5. 글자 색상: color

- #RRGGBB (Red,Green,Blue) 순서로 표현이 되어 있다.
- RGB 색상 값이 두자리 씩 중복이 될 경우 줄여서 표기 가능하다. ex) #0000ff = #00f  
ex) 
② *HSL/ HSLA 표현법
- HSL: Hue(색상) Saturation(채도) Lightness(명도)
- HSLA: HSL + Alpha(불투명도)

  
ex) 
③ *영문 표기
- Red, Yellow, Blue, Black 등 잘 알려진 이름을 사용한다.
- 가장 간편하나, 표현할 수 있는 색상의 수가 제한적인 한계가 있다.  
ex) 
④ *RGB/RGBA
- RGB: Red Green Blue (각각 색상 값은 0~225)
- RGBA: RGB + Alpha

ex) 
** HTML/CSS 색상표: https://htmlcolorcodes.com/
6. 텍스트 정렬: text-align


- start/end는 상대적인 위치를 기준(오-왼쪽으로 쓰는 일부 언어의 경우 다르게 맞춰짐), 좌/우/가/양은 절대적인 위치를 기준으로 한 정렬이다.
7. 글자 간의 간격(줄 간격): line-height

- normal: 웹브라우저에서 정한 기본값이다. 보통 '1.2'이다.
- length(px): 길이로 줄 높이를 정한다.
- number(배수): 글자 크기의 몇 배인지로 줄 높이를 정한다. 
- percentage(%): 글자 크기의 몇 %로 줄 높이로 정한다.
- initial: 기본값으로 설정한다.
- inherit: 부모 요소의 속성값을 상속받는다.

- 이는 글자 크기인 font-size를 기준으로 한다.
8. 텍스트 줄 긋기: text-decoration

- none: 줄 없음
- underline: 밑줄
- overline: 윗줄
- line-through: 취소선
 
9. 그림자 효과: text-shadow

- 가로/세로 거리: 필수 속성 (양수는 아래쪽 그림자, 음수는 위쪽 그림자)
- 번짐 정도: 기본값(0)
- 색상: 기본값(현재 글자색)이 있으나, 한가지 지정/공백으로 구분해서 여러 색상 지정 가능하다.
10. 대/소문자 변환: text-tranformation

- 기본값: none
- capitalize: 맨 앞글자만 대문자
- uppercase: 전부 대문자
- lowercase: 전부 소문자
- full-width: 전각문자로 변환 (가로/세로 비율이 최대한 1:1로 같은 문자)
11. 글자/단어 간격 (자간조절): letter-spacing/word-spacing

12. 기타: 여러 개의 옵션 설정  
① 여러 개의 글꼴 넣기: 지원되지 않는 글꼴을 대비하여 사용
  
  ex) 
② font: 이용한 폰트 스타일 나열

  - 조건1: font-size와 font-family는 무조건 포함해야 한다.
  - 조건2: font-size와 font-family는 반드시 마지막에 와야 한다.
  - 조건3: font-style, font-variant, font-weight는 font-size 앞에 나와야 한다. 
  - 조건4: line-height는 font-size 바로 뒤에 /를 붙여서 나온다. ex) 16px/3
  - font-style, font-variant, font-weight는 순서 상관없이 나열할 수 있다.  
  ex)

```python
# 글꼴 지정
font-family: "글꼴 이름1",이름2,이름3,... # space로 나누어지면 따옴표를 이용하여야 한다.

# font-size
font-size: 절대 크기|상대 크기|크기|백분율

# font-style
font-style: normal|italic|oblique
    
# font-weight
font-weight: normal|bold|bolder|light|lighter|100|...|900
    
# font-color
color: font-color
       hsla(각도,채도(%),명도(%),불투명도(%))
       (rgb(red,green,blue,불투명도(%)))

# text 정렬
text-align: start | end | left | right | center | justify | match-parent
    
# font-variant
font-variant: normal|small-caps|all-small-caps

# line-height: 줄 간격
P { font-size : 12px; line-height : 24px;} # px 절대단위
P { font-size : 12px; line-height : 2.0;} # 배수
P { font-size : 12px; line-height : 200%;} # %

# text-decoration
text-decoration: none|underline|overline|line-through
<p style = “ text-decoration : none”> # 줄 없음
<p style = “ text-decoration : underline”> # 밑줄
<p style = “ text-decoration : overline”> # 윗줄
<p style = “ text-decoration : line-through”> # 취소선
    
# text-shadow
text-shadow: none | <가로거리> <세로거리> < 번짐정도> <색상>
    
# text-transformation
Text-transformation : none | capitalize | uppercase | lowercase | full-width
    
# letter-spacing/word-spacing
letter-spacing: px | em | % 등
```
