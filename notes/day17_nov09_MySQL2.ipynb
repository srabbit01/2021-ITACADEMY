## Stored Procedure (저장 프로시져)  
: SQL문을 하나로 묶어주는 기능  
- 파이썬의 사용자 정의함수와 비슷하다.
- 어떤 동작을 일괄적으로 처리하기 위해 사용된다.

- DELIMITER 는 구분자를 설정해 주는 것인데, myProc()을 묶어주기 위해 임시로 사용했다.
- 이미 생성된 시트는 DROP(삭제)를 해야 재작동이 가능하다. -> 이미 작동한 것은 ERROR 발생 (CREATE는 생성, DROP 은 삭제)
- PROC()은 해당 프로시져를 작동시키는 KEY이며, PROC()에서 (괄호)는 입력할 데이터가 있을 시 입력하는 칸이다.


#### ** DELIMITER
- SQL에서는 ' ; '을 명령 구분자로 사용하는데, ' // '로 임시적으로 바꾸는 것이다. 즉, 구문의 끝을 임의로 설정하는 것.
- 이는 ' ; '을 쿼리 내에 많이 사용할 경우, 한번에 쿼리를 작동시키기 위해 사용한다. (HTML의 DIV와는 다른 개념)

## Trigger (트리거)  
: 테이블에 붙이는 '센서등' (중요한 정보를 담음)  

1. 테이블에 트리거를 붙인다.
2. 테이블에 insert/update/delete 작업을 걸 때
3. 트리거가 함께 실행된다.  
ex) 문제: 사랑이가 회원 탈퇴를 하려고 한다. -> DB에서 사랑이 drop 설정 -> but 회원 중 누가 탈퇴했는지 할 수 없음  
해결: 삭제하기 전에 다른 곳에 복사해 두면 된다. -> but 수작업으로 진행하면 1) 실수/깜빡 발생 2) 귀찮음 -> 따라서, 자동으로 이를 실행하는 것이 Trigger이다.

** NOTE: SQL은 indent sensitive/ case(대소문자) sensitive하지 않다. -> 물론 대문자를 사용하여 구분하면 처리가 조금 빠르다.

** Table realtionship
- 실선(Identifying): 식별 관계. parent table의 primary key(기본키)가 child table의 primary key로 전이되는 식별 관계. 부모가 자식의 모든 정보를 저장하게 된다. (단점으로 작용하는 경우가 많음) child table의 정보가 지워지지 않는한 parent table에 정보가 지워지지 않는다.
- 점선(Non-Identifying): 비식별 관계. child table의 일반 속성 그룹의 구성원으로 전이되는 비식별 관계. 부모가 자식의 정보를 부분적으로만 저장하게 된다. 그러나, 관계가 복잡해질 수록 데이터가 커진다. parent table과 child table의 정보는 각각 독자적인 의미를 가진다.

3. 저장한 모델을 실제 데이터베이스에 적용하기
- MySQL 상단의 File에서 Open_Model을 클릭하기 (저장한 모델 파일 부르기)
- 데이터베이스 -> 모델: Foward Engineer 열기

- 모델 -> 데이터베이스: Reverse Engineer 열기 -> stored procedured와 trigger는 포함되어 나타나지 않는다.
