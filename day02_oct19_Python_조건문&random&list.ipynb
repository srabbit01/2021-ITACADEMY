# If 조건문
- 'If 조건'을 선언해 주고 조건을 넣는다.
- If 조건 선언 후 아래 실행문장 입력란에 indent(띄여진 칸)이 발생하는데, 이는 위계를 의미하며 위의 조건에 종속되었음을 의미한다.

- indent를 실수로 없앴을 때, tab(=space 네번 누르기)를 누르면 하나의 indent를 만들 수 있다.
- 블럭 단위는 움직이자 하는 부분을 누르고 'ctrl + 대괄호'로 움직일 수 있다.

- KEYPOINT: ':'을 넣지 않으면 Error 발생
- 조건이 참일때 실행된다.

- 조건문 연산자

- 조건문 종류
1. 1가지 조건만 다는 경우
2. 조건문 안에 조건문: Nested -> 위계가 존재, 즉 선행 조건이 충족해야 다음 조건 충족 여부 확인 (하위 조건문)
3. if를 연달아 사용하기: ,multi condition -> 동일선상에서 여러가지 조건만 존재
4. 논리연산자:

   and(a and b): a와 b 모두 True여야 True

   or(a or b): a 혹은 b 중 하나만 Ture 여도 True

   not(a is not b): a != b -> a는 b가 아님

- tip
1. 조건이 여러개 있다. (Nested or Multiple)
2. 조건에 위계가 필요한가? Yes = Nested

```python
# model3: Multiple if statement (다중 조건문)
height = float(input("키를 입력하시오: "))
age = float(input("나이를 입력하시오: "))
picture = input("사진은 출력하시겠습니까?(y/n) ")
print()
if height > 120:
    print("입장")
    
    # nested conditional statement -> 하위 조건을 의미
    if age == 19: # 첫 조건만 무조건 "if"
        print("19일이기에 공짜입니다")
    elif age > 18: # 처음과 마지막을 제외한 모든 것은 "elif"
        print("입장료는 5천원입니다")
    elif age == 18:
        print("수험생은 공짜입니다")
    else: # 마지막 조건은 무조건 "else"
         print("입장료는 3천원입니다")
    
    # multiple condition (block 추가) -> 위의 것이 완료되면 실행
    if picture == 'y':
        print("인화료 만원이 추가되었습니다")
    else:
        print("추가되는 금액은 없습니다")

else: # else는 나머지 전부의 경우
    print("입장불가")
```

# List:  list[순서]
- list 목적
1. 많은 값(데이터)을 하나의 변수에 담을 때 필요하다.
2. 데이터의 순서(order)를 반영한다.
- '[ ]'대괄호는 list를 의미한다.
- 빈 list는 우선 띄어두고 나중에 추가해 가면서 사용한다.
- list 순서 -> 순서는 index로 표현이 된다. (주의: 파이썬의 index는 0에서 시작됨)
- list에서 데이터 추가하기
1. 변수.append("넣고 싶은 내용"): 내용을 위에서 부터 나열된다.
2. 변수.extend(['내용1', '내용2', ...])
3. 변수 + ["내용1", "내용2", ...]
4. 변수.insert(데이터를 추가할 위치 뒤, "추가할 내용"): 중간에 데이터 삽입하고자 할 때
- list에서 데이터 제거하기
1. del 변수[삭제하고 싶은 인덱스(위치)]
2. 변수.remove("삭제하고 싶은 내용")
- 정렬
1. 오름차순 정렬: 변수.sort() 또는 변수.sort(reverse=False)
2. 내림차순 정렬: 변수.sort(reverse=Ture) 또는 변수.sort()/변수.reverse()
- 랜덤으로 list 섞기:  
import random  
random.shuffle(list_이름)

```python
# 리스트에 새로운 데이터 추가하기

# append는 각각 list를 나열하는 것이다
list2 = []
list2.append("append로 추가한 첫번째 내용")
list2.append("강남아이티")
list2.append("캪틴아메리카노")

#insert(테이터 추가할 위치를 "인덱스"기준으로, "추가할 내용") -> 지정한 위치 바로 뒤에 배열
list2.insert(2, "인서트로 추가한 내용")

list2
```

```python
# 리스트에서 데이터 제거하기

# del: 삭제하고 싶은 인덱스 입력
print("삭제전 목록: ", list2)
del list2[2]
print("del로 삭제후 목록: ", list2)

# remove: 삭제하고 싶은 내용 그대로 넣기
list2.remove("강남아이티")
print("remove로 삭제후 목록: ", list2)
```

```python
# 오름차순 정렬
list3 = [1,3,4,6,5,9,2,7,8]
list3.sort(reverse=True)
print("정렬 후 list: ", list3)
```

## List 속 요소들을 대/소문자로 모두 바꾸는 방법
1. 새로운 list를 만든다. list = []
2. for i in 해당list:
       새로운list.append(i.upper) -> 요소들을 각각 나누어 준 후 대/소문자로 바꾸고 다시 list로 넣는다.

```python
# 전부 대문자로 만들기
alpha = ['A', 'b', 'C', 'd']
alpha_upper = []
for i in alpha:
    alpha_upper.append(i.upper())
print(alpha_upper)
```

```python
# 전부 소문자로 만들기
alpha = ['A', 'b', 'C', 'd']
alpha_lower = []
for i in alpha:
    alpha_lower.append(i.lower())
print(alpha_lower)
```
