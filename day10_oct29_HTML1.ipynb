# HTML (Hyper Text Markup Language)
- Markup은 '표시함'을 의미. 따라서, 웹문서 사용할 때, 사용하는 언어
- 현재는 version 5를 사용하고 있다.
- HTML은 tag로 이루어져 있다.

ex) 

- HTML 작성 구조: 기본구조는 '!'만 입력하면 자동 완성이 된다.

- 위계는 파이썬과 같이 indent를 이용하여 이루어진다.

```python
<!DOCTYPE html>
<html lang = "ko"> # html이라는 문서 선언
    <head>
        <title>Web Page 이름</title>
    </head>
    <body>
        <h1>소재목</h1>
        <p>내용1</p>
        <p>내용2</p>
    </body>
</html>
```

### Tag 종류
1. Paragraph tag: 문단/단락
2. Anchor tag
3. Semantic(시맨틱) tag:
- 종류:  
① main: 메인 컨텐츠가 들어가는 부분  
② article: 신문/잡지/논문 "기사" 독립된 내용이 나열되어 있을때, 독립된 web contents를 표시한다.  
③ section: 컨텐츠 영영을 표시해준다. Article과 다르게 몇 개의 콘텐츠를 묶는 용도로 많이 사용  
④ aside: 사이드바 영역, 왼/오른/아래쪽에 사이드 바를 만들 수 있다.  
⑤ Footer: 푸터 영역. 맨 아래쪽에 반복되는 내용 -> 회사/사이트에 대한 정보, 연락처, 저작권 등 표기  
- 사용 이유:  
① 코드만 봐도 직관적으로 판단이 가능  
② 구도가 상세하고 정확히 나누어짐 -> 재배열 가능  

4. Div tag**: 
- Semantic tag가 나오기 전에 section들을 구분하기 위해 사용되었다.
- 문서 구조 만들 때 혹은 스타일 적용할 때 사용: 영역을 구별하면서 해당 영역의 스타일까지 지정하는 것
- Id나 class 속성을 이용하여 구조를 만들어 주었다.

- Semantic 대신에 Div를 사용하는 경우, class를 이용하여 작성 목적을 작성해 준다.

5. Text 입력 tag
- <hn'>: 제목. 크고 진하게 표시 -> n = 숫자로 해당 숫자가 증가할 수록 글자 크기가 작아진다. (h1~h6까지 있음)
- <p'>: Paragraph(단락) 텍스트 앞뒤로 빈 줄이 생기면서 단락을 형성한다.
- <br'>: Break Line  
① 편집기에서 줄바꿈을 한다고 해서 줄바꿈이 적용되지 않는다.  
② 줄바꿈을 해준다.
- <blockquote'>: 인용문 블럭. 인용할 때 입력
- <b'>/<strong'>: 글자가 진하게 표기된다. -> 그러나 strong의 경우, 글 자체가 highlight된다.  
① bold: 글자만 진해진다.  
② strong: 글자 자체가 highlight되어 진해진다. -> 경고, 주의사항 표시 등에 사용  
- <i'>/<em'>: Italic 글씨체로 표기된다.  
① i: 이탤릭체로 강조할 텍스트  
② em: 이탤릭체로 표시할 텍스트  
6. Span tag: div와 같이 특별한 기능을 갖고있지 않고, 영역이 아닌 line 내부의 특정 부분을 지정할 때 사용된다.

## 목록 만들기
1. <ol'>, <li'>  
① Ordered List(ol): 순차적으로 나열된 리스트. 리스트를 시작할 때, OL 태그로 먼저 '여기 리스트 시작해요!'라고 선언을 해준다. (OL의 경우 지정하지 않아도 번호가 붙는다.) -> 번호를 붙이고 싶지 않으면 <ul'> 선언  
② List: OL로 선언을 한 다음에 내용을 나열해줄 때, <li'>를 각 내용 앞에 붙여준다.  



2. <ul'>, <li'>  
① Unordered List: 항목에 순서가 중요하지 않을 때 사용한다.  
② List  


3. <dl'>, <dt'>, <dd'>  
① Description List(dl): 선언  
② Description Tag(dt): 키값  
③ Definition Description(dd): 내용  



## 표 만들기1: 표/행(가로)/열(세로)/셀

1. Table tag를 선언해 준다. <table'> </table'>
2. Row로 row 값을 정해준다. <tr'> </tr'>
3. colum으로 행 값을 정해준다.  
① header<th'> </th'>  
② Definition <td'> </td'>  


```python
<table>
    <caption> 표 재목 </caption>
    <tr>
        <th> column1 </th>
        <th> column2 </th>
        <th> column3 </th>
    </tr>
    <tr>
        <td> row1col1 </td>
        <td> row1col2 </td>
        <td> row1col3 </td>
    </tr>
    <tr>
        <td> row2col1 </td>
        <td> row2col2 </td>
        <td> row2col3 </td>
    </tr>
</table>
```
