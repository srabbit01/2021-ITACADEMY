# DBMS (Data Base Management System)
- DB = '데이터의 집합'
- 의미  
1. 대용량 데이터 관련  
2. 여려 명의 유저가 동시에 공유한다.  
- DBMS 특징/조건:  
1. 데이터의 무결성 -> DBMS가 제약 조건을 걸어준다.  
ex) 학생 데이터: 모든 학생은 학번이 있어야 한다. 해당 학번은 고유(unique), 즉 겹치는 학번이 있으면 안된다. 또한, 학번만 알면 학생에 관련된 데이터가 모두 조회가 가능해야 한다.  
2. DB에 들어오는 어떤 데이터든 오류가 있으면 안된다. (필터링을 해야 함)  
3. 데이터의 독립성**: DB 크기/저장소 변경이 되더라도 기존에 작성된 응용 프로그램을 받으면 안된다.  
4. 보안 (무결성 관련): 아무나 데이터에 접근할 수 없다. 오직 데이터를 소유하거나 데이터 접근을 허용받은 이만이 접근이 가능하다. + 접근하는 사용자 계정에 따라 권한이 달라야 한다.
5. 데이터 중복 최소화: 동일한 데이터가 여러 개 중복되어 저장되면 안된다.  
6. 응용 프로그램 제작 및 수정의 용의성  
7. 데이터 안정성 향상: 백업/복원 등등 데이터의 손상 방지
- DBMS의 종류: (개발된 순서대로) 계층형 > 망형 > **관계형** > 객체지향형 > 객체관계형  
1. 계층형 DBMS  
① 장점: 검색이 매우 빠르다.  
② 단점: 구조 변경이 힘들다. 또한, 데이터 접근성이 매우 낮다.  
  
2. 망형 (network)  
① 장점: 1:N 혹은 N:N 관계가 지원되며, 데이터 수출이 증가했다.  
② 단점: 계층보다 더욱 복잡해졌기 때문에 프로그래머가 전체 구조를 모두 이해해야 추론이 가능하다.  
  
3. 관계형  
① Table에 데이터를 저장하는 것이다. 따라서, Table만 잘 이해한다면 데이터 작업을 할 수 있다.  
② 장점: 무결정을 보장해주며, 대용량 데이터 수용이 가능하다.  
③ 단점: 속도가 느리며, 시스템을 resource를 많이 먹는다. -> 현재는 하드웨어가 많이 발달하여 한계 극복이 가능하다.  

# SQL: DBMS를 사용하기 위해 만들어진 언어
- SQL은 DBMS 종류 중 관계형을 이용한다.
- SQL 특징:  
① DBMS 회사와 독립적이다. -> 그러나, 표준성이 공존한다. (즉, 표준은 존재하나 이외에는 회사 별로 특징이 있음)  
② 이식성이 좋다: SQL 표준은 서버/개인/휴대용 장비에서 운영되는 DBMS마다 상호 호환성이 좋다.  
③ SQL 표준은 꾸준하게 연구/업데이트 된다.  
④ 대화식 언어이다. -> 대화식을 칠 때 컴퓨터가 응답하는 것  
⑤ 분산형 구조를 가지고 있다. -> 중앙에 서버가 있고 그것을 거쳐 데이터가 전달이 된다.  
- SQL 한번 배우면 다른 SQL 언어들에서 겹치는 부분이 많고, 표준 SQL은 모두 같다.

### 데이터 타입
1. 문자형 데이터

- CHAR(n): n개의 칸을 지정했을 때, n개의 칸을 모두 사용하지 않더라도 데이터로 입력하여 인식하는 것(데이터 입력 칸을 고정 확보). 데이터 칸이 정해졌기 때문에 처리가 빠른 장점(성능이 좋음)이 있으나, 데이터 용량이 커진다는 단점이 있다. ex) CHAR(10) 3글자만 적어도 10글자 여지 용량 차지
- VARCHAR(n): n개의 칸을 지정했을 때, n개의 칸을 모두 사용하지 않으면 입력된 데이터만큼만 저장하는 것(n은 maximum일 뿐). 데이터 용량에 따라 저장하기 때문에 데이터 용량이 비교적 작은 장점(용량 효율적)이 있으나, 각각의 데이터 칸을 인식해야 하기 때문에 CHAR에 비해 처리 속도가 느린 단점이 있다. ex) VARCHAR(10) 3글자를 입력하면 3글자 만큼만 용량 차지

## View Table (가상 테이블)  

- 가상의 테이블이나, 테이블처럼 취급하여 작동한다. 그러나, 실체가 없어 저장이 되어 있지 않다.
- 어떤 특정 정보를 대량으로 변경할 때, 원본을 모두 수정하는 권한을 갖게 되면 데이터 소실/오류 등의 위험이 있기 때문에 일부 변경 권한만 주고 수정하는 것이다. 그럼에도 원본을 수정하는 것이기 때문에 특정 과정을 거치는 등 데이터 소실/오류 등의 위험에서 최소화 한다.  
ex) 쇼핑몰 운영을하다가 회원주소를 주소를 대량으로 변경해주는 작업이 필요하다. -> 손이 모자라 알바를 고용했다. 하지만 알바에게 데이터 원본을 넘길 수는 없다. -> 왜냐하면 1) 개인정보가 유출이 될 수도 있고. 2)원본데이터가 훼손이 될 수 있기 때문이다.  
그렇다면 선택지는 : 데이터에서 수정이 필요한 부분만 따로 떼어서 알바에게 넘길 수 있다. <- 여기에도 문제가 있다. 바로 DBMS 법칙 중에서 Duplicate 최소화의 법칙을 깬 것이다.  
일단 이미 존재하는 데이터를 새로 생성 하는것도 비효율적이며, 몇 명이 필요할지도 모르는 알바에게 여러개의 원본 duplicate 가 넘어 간다면 어느 db를 기준으로 수정을 적용할 것인가?  
이때 뷰 테이블을 사용한다.  
뷰 테이블은 원본테이블의 데이터를 연결해주는 가상의 테이블이며, 일반적인 테이블처럼 운용이 가능하다.
