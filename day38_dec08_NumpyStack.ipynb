## 1. List vs Array

: 공간을 많이 차지함에도 빠르기 때문에 크기를 고정하여 사용하는 경우가 많다.

```python
import numpy as np
list1 = [1,2,3]
list1.append(4)
list2 = list1 + [5] # 하나의 요소를 추가하는 것이다
print(list2)
list3 = list1 * 2 # 요소를 2배 반복하는 것이다
print(list3)
list4 = [i+3 for i in list1] # 각각의 요소에 숫자를 더하려면 for loop을 사용해야 한다
print(list4)
list5 = [i**2 for i in list1] # 각각의 요소에 제곱근하기 위해서는 for loop을 사용해야 한다
print(list5)
for i in list1:
    print(i)

print()

array1 = np.array([1,2,3])
# numpy에는 append가 없다
a = array1 + [5] # 각각의 요소에 해당 숫자를 더하는 것이 된다
print(a)
b = array1 * 2 # 각 요소에 2배 곱을 하는 것이다
print(b)
for i in array1:
    print(i)
```

```python
a = np.array([1,2])
b = np.array([3,4])

# numpy를 이요하지 않고 내적 구하기
# 첫번째 방법
dot = 0
for e,f in zip(a,b):
    dot += e*f
dot
# 두번째 방법
dot = 0
for i in range(len(a)):
    dot += a[i] * b[i]
dot

# 이는 numpy를 사용하여 여러가지 방법으로 나타낼 수 있다
np.sum(a*b)
(a*b).sum()
np.dot(a,b)
a.dot(b)
b.dot(a)
a@b
```

```python
# 파이썬 기본 사용과 Numpy 사용의 속도 차이
## speed comparison ##
import numpy as np
from datetime import datetime

# note: you can also use %timeit

a = np.random.randn(100)
b = np.random.randn(100)
T = 10000

def slow_dot_product(a, b):
  result = 0
  for e, f in zip(a, b):
    result += e*f
  return result

# 파이썬 기본코드
t0 = datetime.now()
for t in range(T):
  slow_dot_product(a, b)
dt1 = datetime.now() - t0

# 넘파이 내적함수
t0 = datetime.now()
for t in range(T):
  a.dot(b)
dt2 = datetime.now() - t0

print("dt1 / dt2:", dt1.total_seconds() / dt2.total_seconds())
```
