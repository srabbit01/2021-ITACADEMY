### HangMan

1. 랜덤하게 단어 생성: 리스트 단어 3개 정도 random 추출 -> random.choice()
2. 라이프 단어만큼 출력한다. # 라이프는 5개
3. user에게서 입력을 한다.
4. 3의 입력물이 1.의 단어 spelling 중 포함되는지 확인한다. -> for loop을 이용하여 spelling 글자 단위 확인

5. 결과처리
   1) 4.의 결과물이 하나라도 True인 경우 - 맞는 스펠링 표시 (라이프 그대로)
   2) 4.의 결과물이 모두 False인 경우 - 라이프 하나 제거
##### 게임이 끝나는 조건: 라이프가 없거나, spelling을 모두 맞춘 경우

```python
# 입력 only 소문자
fruit = ['apple', 'banana', 'cherry', 'orange', 'melon', 'watermelon', 'lemon', 'mango', 'kiwi']
from string import ascii_lowercase
alpha_list = list(ascii_lowercase)

import random
fruit_ran = random.choice(fruit)
print("%s의 글자가 입력되어야 하며, 당신의 라이프는 5개입니다." %len(fruit_ran))
print()
no1 = len(fruit_ran)
no2 = len(fruit_ran)
life = 5
while not life == 0:
    print(alpha_list)
    word = input("글자 3개를 ,로 나누어 입력하세요: ").split(', ')
    alpha_list.remove(word[0])
    alpha_list.remove(word[1])
    alpha_list.remove(word[2])
    for i in range(0,len(fruit_ran)):
        if word[0] == fruit_ran[i]:
            no1 -= 1
            print("%s가 일치하였습니다."%word[0])
        else:
            pass
        
        if word[1] == fruit_ran[i]:
            no1 -= 1
            print("%s가 일치하였습니다."%word[1])
        else:
            pass
        
        if word[2] == fruit_ran[i]:
            no1 -= 1
            print("%s가 일치하였습니다."%word[2])
        else:
            pass
    if no1 == no2:
        life -= 1
        print("어떤 글자도 일치하지 않기 때문에 라이프가 하나 차감되어 당신의 라이프는 %s입니다." %life)
    print("남은 글자 수는 총 %s개입니다." %no1)
    no2 = no1
    if no1 == 0:
        print("성공하셨습니다! 작성을 완료하신 글자는 %s입니다!" %fruit_ran)
        break
    print()
if life == 0:
    print("BOOM!")
```

```python
def add(n1,n2):
    return n1+n2

def substract(n1,n2):
    return n1-n2

def multiply(n1,n2):
    return n1*n2

def divide(n1,n2):
    return n1/n2
```

```python
# 연산자를 입력 받아서 (+,-,*,/) 위 함수 중. 연산자에 해당되는 사용정의 함수를 실행하라.
# 연산자 입력은 dictionary를 활용하라.
again = 'y'
while not again == 'n':
    a = int(input("첫 번째 숫자를 입력하세요: "))
    calculate = {"+": add, "-": substract, "*": multiply, "/": divide}
    cal_pick = input("연산자를 선택하세요: ")
    b = int(input("두 번째 숫자를 입력하세요: "))
    c = calculate[cal_pick](a,b)
    print(f"현재 값은 {c}입니다.")
    want = input("현재 값으로 계산을 계속하기 원하시면 'y'를 입력하세요: ").lower()
    if want == 'n':
        print(f"최종 계산 값은 {c}입니다.")
    e = 0
    while want == 'y':
        d = int(input("숫자를 입력하세요: "))
        e = calculate[cal_pick](c,d)
        print(f"현재 값은 {e}입니다.")
        want = input("현재 값으로 계산을 계속하기 원하시면 'y'를 입력하세요: ").lower()
    again = input("계산을 처음부터 하고 싶으시면 'y'을 입력하세요: ").lower()
    if again == 'n':
        print(f"최종 계산 값은 {e}입니다.")
    print()
```

```python
# Answer
# 숫자를 두개 입력받아 덧셈/뺄셈/곱셈/나눗셈 해주는 사용자 정의함수를 만드셈

from IPython.display import clear_output # 해당 연산을 초기화하는 것

def add(n1, n2):
    return n1 + n2

def subtract(n1,n2):
    return n1 - n2

def multiply(n1, n2):
    return n1 * n2

def divide(n1,n2):
    return n1/n2

# 연산자를 입력 받아서 (+,-,*,/) 위 함수 중. 연산자에 해당되는 사용정의 함수를 실행하라.
# 사전을 활용해서 키값으로 연산자를 입력 받아  내용인 함수 이름을 출력해준다.
# <- 이 사전을 만들어라. 사전 : "키" : "내용" => 
# 키 = 연산자(+,-,*,/) : 사용자정의함수이름(add,subtract,multiply,divide)
연산자_사전 = {
    "+" : add,
    "-" : subtract,
    "*" : multiply,
    "/" : divide,
}

def 계산기():

    no1 = float(input("첫번째 수를 입력하세요."))
    # 연산자를 입력 받아서 (+,-,*,/) 
    for i in 연산자_사전:
        print(i)

    should_continue = True

    while should_continue: # should_continue는 boolean(true/false)이기 때문에 지정하지 않아도 'true'로 생각

        연산자_입력 = input("연산자를 선택하세요: ")
        no2 = float(input("두번째 수를 입력하세요."))
        #연산자_사전[연산자_입력] # 연산자를 받아서 함수를 '지정'해준다.
        연산_함수 = 연산자_사전[연산자_입력] #지정한 함수 이름을 '연산_함수'에 할당을 해준다.
        # 연산함수의 이름으로 함수를 시행한다.
        answer = 연산_함수(no1,no2)
        print(" %s %s %s = %s" %(no1,연산자_입력,no2,answer)) #결과를 출력해 준다


        # 연산이 끝난 후.
        #계산을 계속할 것인지 사용자에게 물어보고 

        #y 라고 답 하면 계산을 이어서,
        if input(f'''현재 값은 {answer} 입니다. 
    현재 값으로 계산을 계속하기 원하시면 'y'를 입력하세요. 
    계산을 처음부터 하고 싶으시면 'n' 을 입력하세요''').lower() == 'y' :
            no1 = answer
        else: # n 라고 답하면 출력창을 clear 한 다음 계산을 처음 부터 다시 시작하여라.
            should_continue = False
            clear_output() # 완전히 초기화되어 while loop의 처음으로 돌아간다.
            계산기() # 함수 안에서 자체적으로 스스로 다시 실행되도록 설정할 수 있다. <- Recursive 함수 (재귀함수)

계산기()
```
