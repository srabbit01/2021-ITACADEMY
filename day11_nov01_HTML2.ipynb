## 이미지 삽입

- src: 이미지가 저장되어 있는 경로
- alt: 개체 속성이라 하여, 이미지가 읽히지 않을 때 이미지 대신 들어갈 정보(표시)
- 가져올 수 있는 이미지:  
  ① GIF(Graphic interchange format) : 사용 가능한 색상 256개  
  ② JPG/JPEG: GIF보다는 최신이며, 사용 가능한 색상이 더 다양하다.  
  ③ PNG: 위 이미지 파일 중 가장 고화질이다.  
  ④ SPG: 고용량 이미지  
- 현재 작업 경로를 기준으로, 작업 폴더까지 경로를 제외하고 입력한다.

- 사이즈 조절: width, height 로 이미지 사이즈 조절 (둘 중 하나만 지정해도 원본 비율에 따라 자동으로 조절) -> 크기 설정 기준: % or px  
  ① px: 픽셀 단위로 절대적인 크기 설정 (브라우저 크기에 따라 변하지 않음)  
  ② %: 브라우저 너비/높이를 기준으로 % 크기 설정  

```python
<img src="파일_경로" alt="이미지_이름">
```

## 오디오/비디오 삽입  
- <embed'> <object'> : 오디오/ 비디오, PDF, 자바 애플릿 등 멀티미디어 파일 삽입에 사용한다. 
- 비디오 종류:  
  ① mp4: 무료  
  ② webm: 무료이며 화질이 좋다.  
- 오디오 종류:  
  ① mp3: 무료로 전환  
  ② mp4: mp3보다 용량이 작으나 더 좋음  
  ③ mpa  

1. <embed'>

- 옛날식 코드의 삽입 -> 구 버전 explorer 삽입을 위해서는 사용할 수도 있다.
- 이미지와 입력 방식이 완전히 같다. 따라서, closing 태그가 없다.
2. <object'> </object'>

3. <audio'> or <video'>
- 이미지와 입력 방식이 같으나, closing 태그가 필요하다.
- 크기를 미리 정해주는 것이 좋다.
- 비디오를 삽입할 때, 'controls' 속성 값을 같이 넣어주어야 play 조작이 가능하다.

- 자주 쓰이는 태그 속성  
  ① control: 동영상 재생  
  ② autoplay: 페이지 로드 시 자동 재생  
  ③ loop: 반복 재생  
  ④ muted: 소리 제거 (오디오 포함)  
  ⑤ preload: 페이지 로동 시 동영상/오디오 로딩 방싱 정하기 -> auto(기본: 속도 맞춤), metadata, none  
  ⑥ poster = “파일 이름”

```python
<object width=”너비” height =”높이” data=”파일”> </object>

<embed src=”파일경로” width=”너비” height =”높이”>

<audio  src=”파일경로”controls> </audio>

<video src=”파일경로”> </video>
```

# 하이퍼링크: <a'> </a'>

- 자주 쓰이는 태그 속성  
1. href: 연결하고 싶은 주소를 지정  
2. name: 연결하고 싶은 위치의 이름을 지정. HTML5 부터는 사용하지 않음  
3. target: 링크를 클릭했을 때 창의 형태를 지정. '_blank/_self/_top/_bottom'등 여러가지 종류가 있음  
  ① _blank: 새로운 웹 브라우저 창에 하이퍼 링크로 연결한 문서가 나타나게 한다.  
  ② _self: 현재와 같은 웹브라우저 창에 하이퍼링크로 연결한 문서를 보여준다.  
  ③ _top: 프레임 문서의 구조라 할지라도 웹 브라우저 창의 형태로 하이퍼링크로 연결한 문서를 보여준다.  
  ④ _bottom  
  ⑤ _parent: 프레임 문서의 구조에서 하이퍼링크로 연결한 문서를 현재 프레임이 포함된 상위 프레임 창에 나타나게 한다.  
4. title: 링크에 대한 설명이 들어간다


## 폼 삽입하기


- Input의 한 종류
- form의 요소  
  ① fieldset: 하나의 form 안에 여러 구역을 나누어 준다.  
  ② legend: <fieldset'>으로 묶인 그룹의 제목  
  
  ③ label: 입력을 받아오는 form 요소에 이름(라벨)을 붙일 때 사용된다. -> 이때, label은 input과 함께 사용
  
  ④ input 태그: http://jun.hansung.ac.kr/CWP/htmls/HTML%20Input%20Types.html
  

```python
<form action="">
    <fieldset>
        <legend> 상품 선택 </legend>
        <label> 아이디(6자 이상) <input type="text"> </label> # text는 문자열 처리 의미
        <label for="user-id"> 아이디(6자 이상) </label> <input type="text" id="user-id">
    </fieldset>
    <fieldset>
        <legend> 배송 정보 </legend>
    </fieldset>
</form>
```
