실습1) 학생들의 키 데이터를 모아서 평균 키를 계산하시오. (for loop을 사용할 것)

sample data: student_heights = [180, 124, 165, 173, 189, 169, 146]

```python
student_heights = [180, 124, 165, 173, 189, 169, 146]

sum_height = 0
num = 0

for i in student_heights:
    num += 1
    sum_height += i
    print("\n현재는 %s회차 입니다" %num)
    print("현재 누적 학생 키는", sum_height)

print()
print("최종 누적 키는", sum_height)
print("평균키는", sum_height/num)
print("평균키는", sum_height/len(student_heights))
```

```python
snacks = ["새우깡", '감자깡', '고구마깡', "자갈치"]

string_old = input("기존 과자의 이름을 적으시오: ")
string_new = input("넣고자 하는 과자의 이름을 적으시오: ")

# num을 이용하여 계산
num = 0
for i in snacks:
    if i == string_old:
        snacks[num] = string_new
    num += 1
    
# num을 넣지 않는 것
for i in range(0, len(snacks)):
    if snacks[i] == string_old:
        snacks[i] = string_new
    
snacks
```

```python
student_scores = [78, 65, 89, 86, 55, 91, 64, 89]

max = 0

for i in student_scores:
    if i > max: # 각각 값들을 비교하여 구하는 것
        max = i
print(max)

min = 101

for i in student_scores:
    if i < min:
        min = i
print(min)
```

```python
# 6번 문제
# 369 게임기. 1~100의 까지숫자를 생성하고 369 게임 규칙에 따른 결과를 출력하시오
# 규칙1: 3으로 나누어 떨어지면 박수
# 규칙2: 규칙1이 성립되지 않더라도 3/6/9 중에 하나를 포함하면 박수
for i in range(1,101):
    if (i%3) == 0:
        print(i, "박수")
    else :
        i = str(i)
        clap_counter = 0
        for j in range(0,len(i)):
            if i[j] == "3" or i[j] == "6" or i[j] == "9":
                clap_counter += 1
                
        if clap_counter == 0:
            print(i)
        else:
            print(i, "박수"*clap_counter)
```

```python
# 369 게임 - 3/6/9 나온 정도에 따라 박수치기
for i in range(1,700):
    i = str(i)
    clap_counter = 0

    for j in range(0,len(i)):
        if i[j] == "3" or i[j] == "6" or i[j] == "9":
            clap_counter += 1

    if clap_counter == 0:
        i = int(i)
        if (i%3) == 0:
            print(i, "박수: divisible by 3")

        else :
            print(i)
        
    else:
        print(i,"박수   " * clap_counter)  
```
